1. 'f3'컬럼이 gold이면서 'f2'컬럼이 2인 데이터를 구하시오
2. 찾은 데이터에서 f1컬럼의 분산을 구하시오.(반올림 후 소수 둘재 자리가지 구하시오)

```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
```

```python
cond = df['f3'] == 'gold'
cond2 = df['f2'] == 2
df = df[cond & cond2]
```

```python
round(df['f1'].var(),2)
```
