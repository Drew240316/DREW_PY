```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
```

1. f1. subscribed컬럼이 2024 2월인 데이터를 찾으시오
2. 위에서 찾은 데이터 중 f3컬럼이 gold인 데이터의 개수를 구하시오

```python
df['subscribed']= pd.to_datetime(df['subscribed'])
```

```python
df['year'] = df['subscribed'].dt.year
df['month'] = df['subscribed'].dt.month
```

```python
ex = df['year'] == 2024
ex1 = df['month']  == 2
ex2 = df['f3'] == 'gold'
df = df[ex & ex1 & ex2]
print(len(df))
```
