```python
import pandas as pd
df = pd.read_csv('type1_data1.csv')
```

1. 모든 나이에 1을 더하시오
2. 20대의 views평균과 30대의 views평균의 차이의 절댓값을 구하시오. 반올림 후 소수 둘재자리 까지 계산

```python
df['age']= df['age'] + 1
```

```python
cond1 = (df['age'] >= 20) & (df['age'] < 30)
cond2 = (df['age'] >= 30) & (df['age'] < 40)
```

```python
result = abs(df[cond1]['views'].mean() - df[cond2]['views'].mean())
```

```python
round(result,2)
```
