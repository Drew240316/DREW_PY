```python
# 1. while문을 만들어보세요.
# 초기식> 조건식 참/거짓>참이면 반복식/또는 표현식> 거짓이면 출력

i = 0

while i <= 10:
    print(i)
    i += 1
print("end")
```

```python
# 2. while문을 사용하여 10에서 1까지 값이 하나씩 감소하는 값을 만드세요.

i = 10

while i > 0 :
    print(i)
    i -= 1
print(i)
```

```python
# 3. 0에서 값이 시작하여 입력한 값까지 하나씩 출력해주세요. 

i = 0
while i < 10:
    print(i)
    i += 1
print(i)
```

```python
# 4. 생성 범위가 정해졌을 때, 랜덤한 값을 가져와라. 
#    while인 경우, if인 경우 작성! 

import random

i = 0

while i != 6:
    i = random.randint(1,6)
    print(i)

    # while은 계속 1~5까지 무한대로 생성하고 6을 만나면 끝. 
```

```python
# 6. i는 10부터 시작이다. 2를 더해서 출력하고 21을 넘으면 출력하지 말아라.

i = 10

while i < 21:
    print("출력: ",i)
    i += 2
    
```

```python
# 7. 변수 두 개를 다르게 반복하기.
#    정수 2 5,  4 4,   8 3 , 16 2,   32 1  이 각 줄에 출력되게 만드세요.
#    while에 조건식은 두 개 지정하고 두 변수 모두 변화시키세요.
#    i = 2 , j = 5가 초기식입니다.

i = 2
j = 5

while i <= 32 and j >=1  :
    print("i값: ",i,"j값: ",j)
    i *= 2
    j -= 1
```

```python
# 8. 교통카드 잔액 출력하기
#   표준 입력으로 금액(정수)이 입력됩니다. 1회당 요금은 1,350원으로 교통카드를 사용할 때마다 각 줄에 잔액을 출력하는
#   프로그램을 만드세요. (input에서 안내 문자열은 출력하지 않아야 합니다.) 단, 최초 금액은 출력하지 않아야 합니다.
#   그리고 잔액은 음수가 될 수 없으며 잔액이 부족하면 풀력을 끝냅니다.


fare = 1350
value = int(input("금액을 충전하세요: ")) 

while value >= 1350:
    value -= fare
    print("사용금액: 1350원","남은금액: ",value)

if value <= 1349 :
    print("충전해주세요.")
    
    
    
```
