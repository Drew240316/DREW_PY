```python
# 연습 문제 : 변수 두 개를 다르게 반복하기.
# 정수 2 5, 4 4, 8 3 , 16 2, 32 1  이 각 줄에 출력되게 만드세요.
# while에 조건식은 두 개 지정하고 두 변수 모두 변화시키세요.
# i = 2 , j = 5가 초기식입니다.

# 크고 작은 범위를 지정해볼까?

i = 2
j = 5

while i <= 32 or j >= 1 :
    print(i,j)
    i*=2
    j-=1
```

```python
# 8. 교통카드 잔액 출력하기
#   표준 입력으로 금액(정수)이 입력됩니다. 1회당 요금은 1,350원으로 교통카드를 사용할 때마다 각 줄에 잔액을 출력하는
#   프로그램을 만드세요. (input에서 안내 문자열은 출력하지 않아야 합니다.) 단, 최초 금액은 출력하지 않아야 합니다.
#   그리고 잔액은 음수가 될 수 없으며 잔액이 부족하면 풀력을 끝냅니다.


fare = 1350
value = int(input("금액을 충전하세요: ")) 

while value >= 1350:
    value -= fare
    print("사용금액: 1350원","남은금액: ",value)
    
    
```

```python
# while에서 break로 반복문 끝내기
# while 무한 루프에서 숫자를 증가시키다가 변수 i가 100일 때 반복문을 끝내도록 만든다.


i = 0
while True:
    print(i)
    i+=1
    if i == 10:
        break   # : 콜론을 붙이지 않는다.
        
# 초기식 > while문이 항상 참 > 반복한문장/변화식 > 조건이 참이냐 거짓이냐? 참이면 다음코드 아니면 반복
```

```python
# for에서 break로 반복문 끝내기 
# for에 range를 지정해 0부터 30까지 반복한다.

for i in range(50):
    print(i)
    if i == 30:
        break

# for에 range범위의 조건식 > 출력> 조건이 맞는지 안맞는지? 맞으면 break 아니면 for문 반복
```

```python
# for문과 함께 continue로 실행하기 
# 일부 아래 코드를 실행하지 않고 건너뛰기.

for i in range(10):
    if i % 2 == 0: # 짝수면..
        continue  #아래 코드를 실행하지 않고 건너뜀.
    print(i)
    
# for의 range에 꺼낼 숫자가 있나? 짝수면? continue 홀수면? i출력 
```

```python
# while문과 함께 continue 실행하기
i = 0
while i < 20 :
    i += 1
    if i % 2 == 0:
        continue
    print(i)
    
    
# 초기식 > while의 조건식 > 반복할 문장/변화식 ? 조건이 맞으면 countinue > 아니면 출력 
```

```python
# 입력한 횟수대로 반복하기. 

count = int(input('반복할 횟수를 입력하세요: '))

i = 0
while True :
    print(i)
    i += 1
    if i == count :
        break 
```

```python
# 중첩 루프를 사용해서 바깥쪽, 안쪽 숫자를 출력하고 싶다. 세로 코드가 1번일 때 가로코드는 0~4까지 숫자가 늘어난다.
# 이를 출력하여라. 

for i in range(5):
    for j in range(5):
        print("j는 ", j, "이고, ", sep = ' ', end = ' ') #공백으로 띄우고, end로 줄바꿈 대신 한 칸 띄움.
        
```

```python
# 사각형으로 별 출력하기.
# 중첩된 반복문을 사용하여 5 * 5사각형 형태로 별 그리기

for i in range(5):
    for j in range(5):
        print('*', end = ' ') 
    print()                      # print()안에 공백이면 줄 바꿈을 의미한다.

    
```

```python
# 계단식으로 별 출력하기 - 세로 몇개? 가로 몇 개?

for i in range(5):
    for j in range(5):
        if j <= i :               #세로 방향 변수 i 만큼
            print('*', end = ' ') #별 출력, end에 ' '를 지정하여 줄바꿈을 하지 않음
    print()
```

```python
# 대각선으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j == i :
            print('*', end = ' ')
        else:
            print(' ', end = ' ')
    print()
```

```python
# 역삼각형으로 별 출력하기 - 세로 몇개? 가로 몇 개?

for i in range(5):
    for j in range(5):
        if i <= j :               #세로 방향 변수 i 만큼
            print('*', end = ' ') #별 출력, end에 ' '를 지정하여 줄바꿈을 하지 않음
    print()
```
