# 회문 판별하기

##### 빅데이터 분석, 검색 엔진에서 많이 사용. N-gram viewer은 사람들의 언어 패턴 분석 
##### 회문은 순서를 거꾸로 읽어도 제대로 읽은 것과 같은 단어와 문장을 말한다. sos level rotator 

```python
# 반복문으로 문자 검사하기 

word = input('단어를 입력하세요: ') # 회문 판별값을 저장할 변수, 초깃값은 True 
is_palindrome = True
for i in range(len(word) // 2 ): # 0부터 [문자_서식_포맷_정규] 길이의 절반만큼 반복하여 가운데 글자 바로 앞까지만 검사
    if word[i] != word[-1 -i]:   # 왼쪽 문자와 오른쪽 문자를 비교하여 문자가 다르면 
        si_palindrome = False    # 회문이 아님.
        break 

print(is_palindrome)              # 회문 판별값 출력  
```

```python
# 시퀀스 뒤집기

word = input("단어를 입력하세요:")

print(word == word[::-1])
```

```python
# 리스트와 reversed 사용해서 뒤집기

word = 'level'

list(word) == list(reversed(word))
```

```python
# 문자열의 join 메서드와 reversd 사용해서 뒤집기

word = 'level'
word == ''.join(reversed(word))
```

```python
# 반복문으로 N-GRAM 출력하기
## n-gram은 문자열에서 n개의 연속된 요소를 추출하는 방법입니다.
## 문자열의 처음부터 [문자_서식_포맷_정규] 끝까지 한 글자씩 이동하면서 2글자를 추출합니다.

text = 'Hello'

# 2-gram이므로 문자열의 끝에서 한 글자 앞까지만 반복한다. #i = 4 → text[4] + text[5] → 오류 발생!!
for i in range(len(text)-1):
    print(text[i],text[i+1],sep = '')  #현재 문자와 그 다음 문자 출력 
    
    
# 만약 3-gram이면 반복 횟수는 (range(len(text)-2)) 와 같이 되고, [문자_서식_포맷_정규] 끝에서 두 글자 앞까지 반복하면 됩니다.
```

```python
# zip으로 2-gram만들기

text = 'Hello'

two_gram = zip(text, text[1:])  
for i in two_gram:
    print(i[0],i[1], sep = '') 
```

```python
# zip과 리스트 표현식으로 N-gram으로 만들기
## Ngram을 만들 때 zip에 일일이 [1:][2:]같은 슬라이스를 넣으면 상당히 번거롭습니다.
# 만약 Ngram의 숫자가 늘어나면 그만큼 스랄이스도 여러 개 입력해줘야 합니다
# 그러면 이 과정을 "리스트 표현식"을 사용하면 됩니다.
# 아래는 회문 전 준비 단계 

text = 'hello'
[text[i:] for i in range(3)]
```

표현	 의미

range(N)	N-gram을 만들기 위해 슬라이싱할 횟수

text[i:] for i in range(N)	i번째 위치부터 시작하는 N개의 슬라이스

zip(...)	N개의 슬라이스에서 같은 위치 문자를 묶어서 N-gram 생성

```python
# 이제 리스트 ['hello', 'ello', 'llo']를 zip에 넣습니다.

list(zip(['hello', 'ello', 'llo']))
```

```python
# 3-gram이 아닙니다. zip은 바복 가능한 객체 여러 개를 콤마로 구분해서 넣어줘야 합니다. 
# 하지만 ['hello', 'ello', 'llo'] 의 요소가 3개 들어있는 리스트는 1개 입니다.
# 각 요소를 콤마로 구분해서 넣어주려면 리스트 앞에 *(리스트 언패킹)을 합니다.
list(zip(*['hello', 'ello', 'llo']))
```
