#### 예외 처리 왜 쓰지?

    - 예외 처리는 코드가 멈출지 계속 진행할지 선택하는 제어 장치이다.
        - 분석 자동화 스크립트일 경우, 파일 하나 없다고 전체 멈추면 곤란함
        - 현업에서 CSV/엑셀 데이터는 정말 들쑥날쑥하게 오기 때문에 컬럼이 없을 수도 있음.
        - 데이터 전처리 누락으로 차원이 안 맞거나, 타겟에 NaN이 껴있으면 바로 에러남.
          → 오류를 보여주고 로그 남긴 뒤, 전처리 다시 하게 유도.
        - 실시간 데이터 수집 자동화할 때, 일시적 오류로 분석 전체가 멈추지 않도록.
        - categorical 값이 너무 적거나, 숫자형인데 object 타입일 때 에러남.
          → 분석 보고서 자동화 시 꼭 처리 필요.
          
          
     try:
         실행할 코드
     except :
         예외가 발생했을 때 처리하는 코드 

```python
try :
    x = int(input('나눌 숫자를 입력하세요: '))
    y = 10 / x  # 예외가 발생했을 때 실행되지 않음
    print(x)   # 예외가 발생했을 때 실행되지 않음

except : # 예외가 발생했을 때 실행됨
    print('예외가 발생했습니다.')
    
    
    # 0 을 입력해서 일부로 예외를 발생시켜주세요.
```

```python
# 입력을 2 0으로 해보자
# 입력을 3 5로 해보자

y = [10,20,30]

try :
    index, x = map(int, input('인덱스와 나눌 숫자를 입력하세요: ').split())
    print(y[index].x) 
    
except ZeroDivisionError: #숫자를 0으로 나눠서 에러가 발생했을 때 실행
    print('숫자를 0으로 나눌수 없습니다.')  
    
except IndexError : # 범위를 벗어난 인덱스에 접근하여 에러가 발생
    print('잘못된 인덱스입니다.')
```

#### 예외의 에러 메세지 받아오기 

    try :
        실행할 코드

    except 예외 as 변수:
        예외가 발생했을 때 처리하는 코드 

```python
# 입력을 2 0으로 해보자
# 입력을 3 5로 해보자

y = [10,20,30]

try :
    index, x = map(int, input('인덱스와 나눌 숫자를 입력하세요: ').split())
    print(y[index].x) 
    
except ZeroDivisionError as e : # as 뒤에 변수를 지정하면 에러를 받아옴
    print('숫자를 0으로 나눌수 없습니다.',e)  # e에 지정된 에러 메세지 출력
    
except IndexError as e :
    print('잘못된 인덱스입니다.',e)
```

#### 예외 계층

BaseException
├── SystemExit                # sys.exit() 호출 시 발생
├── KeyboardInterrupt         # Ctrl + C 입력 시 발생
├── GeneratorExit             # 제너레이터 종료 시 발생
└── Exception                 # 대부분의 예외는 여기서 파생
    ├── ArithmeticError       # 산술 계산 오류
    │   ├── ZeroDivisionError     # 0으로 나눌 때
    │   ├── OverflowError         # 너무 큰 숫자 계산 시
    │   └── FloatingPointError    # 부동 소수점 계산 오류
    ├── LookupError           # 잘못된 키 또는 인덱스
    │   ├── IndexError            # 리스트 인덱스 범위 초과
    │   └── KeyError              # 딕셔너리 키 오류
    ├── ValueError            # 값의 타입은 맞지만 부적절할 때
    │   └── UnicodeError          # 유니코드 관련 오류
    ├── TypeError             # 타입이 맞지 않을 때
    ├── NameError             # 존재하지 않는 변수 참조
    │   └── UnboundLocalError    # 지역 변수 참조 오류
    ├── AttributeError        # 존재하지 않는 속성 접근
    ├── ImportError           # 모듈 import 실패
    │   └── ModuleNotFoundError  # 모듈을 찾을 수 없을 때
    ├── FileNotFoundError     # 파일을 찾을 수 없을 때
    ├── OSError               # 운영체제 관련 오류
    ├── RuntimeError          # 실행 중 알 수 없는 오류
    └── StopIteration         # 반복자에서 더 이상 값이 없을 때

🔍 현실 상황 2: 전처리 함수 재사용 시

def normalize_column(df, col):
    if col not in df.columns:
        raise KeyError(f"{col} 컬럼이 없습니다.")
    df[col] = (df[col] - df[col].mean()) / df[col].std()
    return df
자동화된 전처리 스크립트 짤 때 자주 쓰임

필수 컬럼이 없을 때, 애매하게 넘기지 말고 raise로 중단
