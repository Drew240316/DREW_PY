```python
# 클로저 (닫는 사람)
## 클로저는 함수를 둘러싼 환경(지역 변수, 코드 등)를 계속 유지하다가 함수를 호출할 때 다시 꺼내서
## 사용하는 함수를 뜻합니다.
## 따라서 클로저는 지역 변수와 코드를 묶어서 사용하고 싶을 때 활용합니다.
## 프로그램의 흐름을 변수에 저장 
## 또한, 클로저에 속한 지역 변수는 바깥에서 직접적으로 접근할 수 없으므로 데이터를 숨기고 싶을 때 활용
## 데이터 은닉(정보 보호)  API 키 숨기기, 호출 횟수 제한, 함수형 클래스처럼 쓰기 (캡슐화)


# 전역변수 (global variable)
## 전역 변수는 함수 바깥에서 선언되며,
## 함수 안에서는 보통 읽기만 가능하다.

## 하지만 함수 안에서 전역 변수의 값을 변경하려면,
## global 전역변수명을 사용해 전역 변역변수를 가리키겠다고 선언.(덮어쓰기x)


x = 10        # 전역 변수 
def foo():
    print(x)  # 전역 변수 출력, 안에서 읽기 가능 

foo()
print(x)  # 전역 변수 출력, 바깥에서 읽기 가능
```

```python
# 지역변수
## 특정 함수 안에서만 유효한 변수로, 함수 바깥에서는 직접 접근하거나 사용할 수 없다.
## 지역 변수는 함수가 끝나면 사라지는, 함수 내부 전용 변수!
## 지역변수 error 
def foo():
    x = 10
    print(x)
    
foo()
print(x)   # 함수안에서만 접근 가능하고, 바깥에서는 접근 불가
# name 'x' is not defined
```

```python
# 함수 안에서 전역 변수 변경 전!  
x = 10
def foo():
    x = 20
    print(x)   
foo()         # 지역변수 호출 
print(x)      # 바깥에서 호출, 전역변수 물러옴
```

```python
# 함수 안에서 전역 변수 변경 후

x = 10  # 전역 변수 정의

def foo():
    global x           # global**은 바깥의 전역 변수를 가리키기 위해 선언하며,
                         # 해당 변수에 접근하고 값을 변경할 수 있게 해준다.
    print(x)        
foo()                 # 함수 호출: 전역 x(10) 출력
print(x)              # 전역 변수 x 그대로 출력 (아직 값 안 바뀜)
```

```python
# 함수 안에 함수 만들기

전역변수
def 함수이름():
    코드         # 지역 변수 자리 및 코드
                 # 완전 바깥 전역변수 사용하려면 golbal 사용 
    def 함수이름():  
        코드      #지역 변수 자리 및 코드
                  # 변경하려면 nonlocal
```

```python
# 지역변수를 변경하기 : nonlocal (비국소적)
# nonlocal x는 새로운 지역변수를 만들지 않고,
# 바깥 함수에 이미 있던 x (10) 를 그대로 가져와서 그 값을 수정한다. 

def A():
    x = 10 # A의 지역변수.
    def B():
        nonlocal x   # 바깥 함수의 지역변수를 가리키고자 선언하여 그 값을 할당(=수정)하고자 할 때 선언
        x = 20       
    
    B()
    print(x)

A()
```

```python
# global은 바깥의 전역 변수를 사용하고 수정하기 위해 선언한다.
# nonlocal은 바깥 함수의 지역 변수를 가리키고,
# 그 값을 할당(=수정)하고자 할 때 선언한다.
```

```python
# nonlocal이 변수를 찾는 순서

def A():
    x = 10
    y = 100
    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x = x + 30  # x 10
            y = y + 300  # y 100
            print(x)
            print(y)
        C()
    B()
A()

# C B A 순서로 작동, C()에서 B()의 Y가 없으면 A에서 가져옴 
            
```

```python
# global로 전역 변수 사용하기

x = 1
def A():
    x = 10
    def B():
        x = 20
        def C():
            global x
            x = x + 30
            print(x)
        C()
    B()
A()

# C B A 순서로 작동, global쓰면 무조건 전역 변수 1 사용.
```

```python
# 클로저 사용하기 = 프로그램의 흐름을 변수에 저장

def cal(): 
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b # 함수 바깥쪽에 있는 지역변수 a,b를 사용하여 계산
    return mul_add       # 함수를 리턴하되, 괄호는 쓰면안됨!

c = cal()                # 변수c에 저장된 함수가 클로저가 됨. 
print(c(1),c(2),c(3),c(4),c(5))



# 개념	     설명
# x는        함수의 매개변수(파라미터) 이름	def mul_add(x): 안에서만 의미 있음
# c = cal()	c는 이제 cal()이 mul_add 함수가 반환한 값을 가리킴. 그래서c = cal()은 mul_add함수 자체를c에넣음
# c(1)	    c(1) == mul_add(1)  # 같은 의미!


#호출	계산식	결과
#c(1)	3*1 + 5	8
#c(2)	3*2 + 5	11
#c(3)	3*3 + 5	14
#c(4)	3*4 + 5	17
#c(5)	3*5 + 5	20
```

```python
# lambda로 클로저 만들기

def calc():
    a = 3
    b = 5
    return lambda x : a * x + b #람다 표현식 반환

c = cal()

print(c(1),c(2),c(3),c(4),c(5)) 
```

```python
# 클로저의 지역 변수 변경하기

def cal():
    a = 3
    b = 5
    total = 0
    def mul_add(x):
        nonlocal total
        total = total + a * x + b
        print(total)
    return mul_add

c = cal()
c(1)
c(2)
c(3)
```
