```python
# 반복문으로 리스트의 요소를 모두 출력하기.
# 리스트의 인덱스를 지정해서 하나씩 출력하지 않고, 리스트와 반복문을 사용해서 간단하게 모든 요소 출력
# for 변수 in 리스트 :
   #반복할 코드
```

```python
a = [38,21,53,62,19]
for i in a:
    print(i)
# 리스트 a에서 요소를 꺼내서 i에 저장하고, 꺼낼 때마다 코드를 반복
```

```python
# 인덱스와 요소를 모두 함께 출력하려면?
# for 인덱스, 요소 in enumerate(리스트):

a =  [38,21,53,62,19]

for index, value in enumerate(a):  #이뉴멀레이트 : 열거하다
    print(index, value)
    
```

```python
# 리스트 요소에 1부터 번호 매기며 출력하기 

a =  [38,21,53,62,19]

for index, value in enumerate(a):
    print(index+1,value)
```

```python
# 특정 인덱스와 요소를 함께 출력할 때 좀 더 파이썬 스럽게 출력하려면?
# for index, value  in enumerate(리스트, start = 숫자):

a =  [38,21,53,62,19]

for index, value in enumerate(a, start = 2):
    print(index+1, value)
```

```python
# while 반복문으로 요소 출력하기

a =  [38,21,53,62,19]

i = 0
while i < len(a):
    print(a[i])
    i +=1
```

```python
# while 반복문으로 리스트의 요소를 출력할 때는 변수 i를 인덱스로 활용한다.
# 먼저 변수 i를 0으로 만들어주고, i <len(a)처럼 i가 리스트의 길이(요소 개수) 직전까지만 반복하도록 만든다.
# 즉, 리스트의 인덱스는 0부터 시작하고, 마지막 인덱스는 리스트의 길이보다 1이 작으므로 <를 사용한다. 
# <=를 사용하게되면 인덱스의 범위를 벗어나므로 주의하자. 
```

```python
# 가장 작은 수와 가장 큰 수 구하기
# for문을 사용할까요? sort를 사용할까요? min max를 사용할까요?
# 먼저 리스트 a의 첫 번째 요소 a[0]을 변수 smallest에 저장합니다. 
# i가 smallest보다 작으면 smallest에 i를 할당합니다.
# 즉, 숫자를 계속 "비교"해서 숫자가 작으면 smallest를 바꾸는 방식입니다.
a =  [38,21,53,62,19]
smallest = a[0]

for i in a :
    if i < smallest :
        smallest = i
        
smallest
```

```python
# for문을 사용해서 int로 바꾸기
a = [ 1.2, 2.5, 3.7 , 4.6 ]

for i in range(len(a)):
    a[i]  = int(a[i])
    
a
```

```python
# for에 range(len(a))을 사용해서 인덱스를 가져왔습니다.
## a로 접근하면 float형태라서 error남. 인덱스로 접근해야함. 
### list indices must be integers or slices, not float
# 그리고 가져온 인덱스로 하나하나에 접근한 뒤 int로 변환하여 다시 저장했습니다.
# 매번 for 반복문으로 반복하면서 요소를 변환하려니 조금 번거롭습니다. 이럴 때는 map을 사용합니다. 
```

```python
# map은 리스트의 요소를 지정된 함수로 처리해주는 함수이다. (05. 공통_[출력_반복문]★ 리스트를 변경하지 않고 새 리스트를 생성)
# list(map(함수, 리스트))
# tuple(map(함수, 튜플))
```

```python
#map을 사용해서 int로 바꿔보자. 

a = [ 1.2, 2.5, 3.7, 4.6 ]
a = list(map(int,a))
a
```

```python
# 결과가 왜 이럴까?
# map에 int와 같은 리스트를 넣으면 리스트의 모든 요소를 int를 사용해서 변환합니다.
# 그 다음 list를 사용해서 map의 결과를 리스트로 만들어줍니다.
# 사실, map은 리스트뿐만 아니라 반복 가능한 객체를 넣을 수 있습니다. ex. range 
```

```python
a = list(map(str, range(10)))
a
```
