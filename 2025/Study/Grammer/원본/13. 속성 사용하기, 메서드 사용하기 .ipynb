### 1. 클래스 속성 사용하기 : 클래스의 모든 인스턴스 공유

     def 클래스 이름:
         속성 = 값
   

```python
class Person :
    bag = []
    
    def put_bag(self, stuff):
        self.bag.append(stuff)
    
james = Person()
james.put_bag('책')   # bag 항목에 리스트 객체가 할당 o

maria = Person()
maria.put_bag('열쇠') # bag 항목에 리스트 객체가 할당 o 

print(james.bag)
print(maria.bag)
```

 왜 ? james maria 둘다 [책,열쇠]가 나왔을까?
     클래스 속성은 클래스에 속해 있으면 **모든 인스턴스** 를 공유합니다.
     
     그렇다면, 더욱 명확하게 클래스의 인스턴스임을 알게 할 수 없나요?

### 3. 클래스 속성 사용하기 : 여러 인스턴스와 공유하고 싶지 않을 때 사용하기(인스턴스 각자 출력하기)

- 여러 사람(인스턴스)와 공유하지 않고 싶을 때 사용한다. 즉, 인스턴스 각자 출력된다.


- 이유는 바로 __init__() 안에서 self.bag = []를 하기 때문!
   __init__()은 인스턴스를 생성할 때마다 자동 실행된다.
  self.bag = []는 새로운 빈 리스트를 만든 다음,
  그걸 james.bag 또는 maria.bag에 각자 따로 저장한다. 결국 각 인스턴스마다 새로운 리스트(메모리 공간)이 만들   어진다.

```python
class Person:
    def __init__(self):
        self.bag = []
    
    def put_bag(self, stuff):
        self.bag.append(stuff)
        
james = Person()
james.put_bag('책')

maria = Person()
maria.put_bag('열쇠')

print(james.bag)
print(maria.bag)
```

### 4. 비공개 클래스 속성 
비공개 인스턴스 속성은 클래스 내부에서만 사용  

     def 클래스 이름:
         def __init__(self):
             self.__속성 == 값

```python
class Knight:
    __item_limit = 10  #비공개 클래스 속성
    
    def print_item_limit(self):
        print(Knight.__item_limit)   #클래스 안에서만 접근
    
x = Knight()
x.print_item_limit()  # 10
```

```python
print(Knight.__item__limit) # 클래스 바깥에서 접근 불가

# error : 타입 객체 knight는 __item __limit__을가지고있지않아요
```

### 5. 정적 메서드 사용하기.

- 인스턴스(self, 인스턴스 인자)와 상관없이 동작하는 메서드로 클래스에서 바로 메소드 호출(클래스 안에 있는 일반 함수)
- 인스턴스 속성, 인스턴스 메서드가 필요없을 때 사용
- 무엇을 정적 메서드로 만드는가?
  - 메서드의 실행이 외부 상태에 영향을 끼치지 않는 순수 함수를 만들 때
  - 순수함수는 부수 효과(side effect)가 없고 입력 값이 같으면 언제나 같은 출력 값 반환
  - 즉, 클래스 안에는 있지만 인스턴스 상태 = 객체의 상태랑 관련 없는 기능을 넣고 싶을 때 사용

        class 클래스이름:
        @staticmethod
        def 메서드(매개변수1, 매개변수2):
            코드


class Calc:
    @staticmethod
    def add(a,b):
        print(a+b)
    
    @staticmethod
    def mul(a,b):
        print(a * b)
    
Calc.add(10,20) # 클래스에서 바로 메소드 호출
Calc.mul(10,20) # 클래스에서 바로 메소드호출

### 6. 클래스 메서드 

- 클래스 자체에 관련된 기능 구현	 (예: 전체 객체 수 세기, 클래스 이름 출력 등)
- 인스턴스를 만들기 위한 팩토리 메서드 만들 때	(예: 반복적인 객체 생성 과정을 깔끔하게 숨기고 싶을 때, 외부에서 API로 데이터 가져올 때, Person처럼 명확한 목적을 가진 클래스를 생성할 때)
- 공통 설정/환경 초기화	(예: cls.setting = ...) 

         class 클래스 이름:
         @classmethod
         def 메서드(cls, 매개변수1, 매개변수2):
             코드
 - 클래스 속성 ( 공용 속성, __init__없어도 됨)
        
         class Person:
             count = 0  # 클래스 속성 (공통)

         print(Person.count)  # → 0

 - 인스턴스 속성 ( 개별 속성, __init__ 안에서 만들어짐)
   : 인스턴스가 생성될 때마다 개별 데이터를 넣고 싶을 때 쓰는 것! 
   
        class Person:
            def __init__(self, name):
                self.name = name  # 인스턴스 속성

        p1 = Person("민지")
        print(p1.name)       # → "민지"


```python
class Person:
    count = 0 # 클래스의 속성 
    
    def __init__(self):
        Person.count += 1 # 인스턴스가 만들어질 때, 클래스 속성 count에 +1을 함
    
    @classmethod
    def print_count(cls): # self 대신 cls씀 
        print('{0}명 생성되었습니다.'.format(cls.count)) #cls속성에접근

james = Person()
maria = Person()

Person.print_count()   # 2명 생성
```
