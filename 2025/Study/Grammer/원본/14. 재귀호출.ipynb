```python
# 재귀 호출은?
# 함수 자기 자신을 호출하는 방식을 재귀호출이라고 하며
# 일반적인 상황보다도 알고리즘에 ㄸ라서 반복문으로 구현한 코드보다도 재귀호출을 구현한 코드가
# 더 깔끔함 (종읿접기 접기 접기 접기 펴기펴기펴기,,, 백트래킹, 트리 ,dfs..)
# 재귀는 문제를 작게 쪼개서 해결하고,
# 해결한 걸 다시 조립하기 위해 '돌아오는 구조'를 활용하는 것
```

```python
# 재귀호출에는 종료 조건을 만들어야한다.

def hello(count):
    if count == 0:   # 종료 조건 
        return 
    print('Hello world',count)
    
    count -= 1
    hello(count)

hello(5)


# 종료 조건 설정안하면 RecursionError 가 뜬다. 재귀호출이 종료되지 않으면 최대 재귀 깊이를 초과하여
# 에러 발생 파이썬은 1000까지 깊음.
```

```python
# 재귀호출로 팩토리얼 
# 1부터 n까지의 양의 정수르 차례대로 곱한 값이며 !(느낌표) 기호로 표기
# 5! 5*4*3*2*1

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
    
print(factorial(5))
```
