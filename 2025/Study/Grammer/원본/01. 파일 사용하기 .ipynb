# 파일에 문자열 쓰기 읽기 
###  파일에 문자열을 써서 파일을 만든 뒤에 만든 파일을 읽는 순서로 진행 

```python

# 파일을 쓰기 모드로 열기
## 파일 객체 = open(파일이름,파일모드)
file = open('hello.txt','w') # hello.txt 파일을 쓰기 모드(w)로 열기, 파일 객체 반환
```

```python
#파일에 [문자_서식_포맷_정규] 저장
## 파일객체.write('[문자_서식_포맷_정규]')
file.write('Hello,world')   
```

```python
# 파일 객체 닫기 
## 파일객체.close()
file.close() 
```

```python
# 파일을 읽기 모드로 읽기
# 변수 = 파일객체.read()
file = open('hello.txt','r')  # 1. 파일을 읽기모드로 읽고, 파일 객체 반환 
s = file.read()               # 2. 파일에서 [문자_서식_포맷_정규] 읽기
print(s)                      # Hello.Wolrd! 
file.colse()                  # 3. 파일 객체 닫기 

```

```python
# 자동으로 파일 객체 닫기
# with open(파일이름, 파일모드) as 파일객체:
#    코드


with open('hello.txt','r') as file :
    s = file.read()
    print(s)
```

```python
# [문자_서식_포맷_정규] 여러 줄을 파일에 쓰기, 읽기

# 반복문으로 [문자_서식_포맷_정규] 여러 줄을 파일에 쓰기

with open('hello.txt','w') as file :
    for i in range(3):
        file.write('Hello, world! {0}\n'.format(i))   #[문자_서식_포맷_정규] 끝에 {0}\n을 넣어야 줄바꿈이 됨.
```

```python
# 리스트에 들어있는 문자열을 파일에 쓰기 
# 파일객체.writelines(문자열리스트)

lines = ['안녕하세요\n','파이썬\n','입니다.\n']

with open('hello.txt','w') as file: # hello.txt 파일을 쓰기 모드로 열기 
    file.writelines(lines)
```

```python
# 파일의 내용을 한 줄씩 읽기
# 변수 = 파일객체.readline()   # readline은 while을 쓴다. 

with open('hello.txt','r') as file: # hello.txt을 읽기 모드로 읽기
    line = None                     # 변수 line을 None으로 초기화, ''으로 하면 while의 조건이 거짓이 되므로 반복x
    while line != '':               # readline은 더 이상 읽을 줄이 없을 때는 빈 문자열을 반환하는데, while에는 이런 특성을 이용하여 조건식을 만든다.
        line = file.readline()      # 즉, line != ''과 같이 빈 문자열이 아닐 때 계속 반복
        print(line.strip('\n'))     # 파일에서 읽어온 문자열에서 \n 삭제하여 출력 
```

```python
# for 반복문으로 파일의 내용을 줄 단위로 읽기

with open('hello.txt','r') as file:  # hello.txt 파일을 읽기 모드(r)로 열기 
    for line in file:                # for에 파일 객체를 지정하면 파일의 내용을 한 줄씩 읽어 변수에 저장
        print(line.strip('\n'))      # 참고로 file 객체는 이터레이터이다. 
```

# 파이썬 객체를 파일에 저장하기, 가져오기

### 파이썬 객체를 파일에 저장하는 과정을 피클링이라고 하고, 읽어오는 과정을 언피클링이라고 한다. 

```python
# 파이썬 객체를 파일에 저장하기

import pickle

name = 'james'
age = 17
address = '서울시 동작구 동동동'
scores = {'korean':90,'english':95,'mathematics':85,'science':82}
```

```python
with open ('janmes.p','wb')  as file : # hello.txt파일을 바이너리 쓰기 모드로 열기
    pickle.dump(name,file)
    pickle.dump(age,file)
    pickle.dump(address,file)
    pickle.dump(scores, file)
    
# 소스코드를 실행하면, .py파일이 있는 폴더에 james.p파일이 생성됩니다.
# 여기서는 확자자를 pickle의 p를 사용했지만 다른 확장자를 사용해도 상관 없습니다.
# 특히 pickle.dump로 객체(값)를 지정할 때는 open('james.p','wb')와 같이 파일 모드를 'wb'로 지정해야 합니다
# b는 바이너리를 뜻하는데, 바이너리 파일은 컴퓨터가 처리하는 파일 형식입니다. 
# 따라서 메모장 같은 텍스트 편집기로 열어도 사람이 알아보기 어렵습니다. 
```

```python
# 파일에서 파이썬 객체 읽기

import pickle

with open('janmes.p','rb') as file: # hello.txt 파일을 바이너리 읽기 모드(rb)로 열기
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)
    print(name)
    print(age)
    print(address)
    print(scores)
    
```
