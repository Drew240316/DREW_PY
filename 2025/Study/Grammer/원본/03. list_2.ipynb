```python
#반복문으로 2차원 리스트의 요소를 모두 출력하기
# 1. for 바복문을 한 번만 사용하는 방식이다. 
a = [[10,20],[30,40],[50,60]]  
for x, y in a:              # 리스트의 가로 한 줄(안쪽 리스트)에서 요소 두 개를 꺼냄.
    print(x,y)              # 2차원 리스트에 for문을 사용하면 가로 한 줄씩 반복하게 된다. 
                            # 전체 리스트를 기준으로 보면 안쪽 리스트가 통째로 반복된다.
                            # 이 때, for x,y in a :와 같이 in 앞에 변수를 두 개 지정해주면 가로 한 줄(안쪽 리스트)에서
                            # 두 개를 꺼내 온다.
```

```python
# 2. for 반복문을 두 번 사용하기

a = [ [10,20],[30,40],[50,60] ]

for i in a:                        # a에서 안쪽 리스트를 꺼냄 [10,20]
    for j in i:                    # 안쪽 리스트에서 요소를 하나씩 꺼냄 10...20...
        print(j, end = ' ')
```

```python
# 3. for와 range 사용하기 
# for range에 세로 크기와 가로 크기를 지정해서 2차원 리스트의 요소를 인덱스로 접근

a = [ [10,20],[30,40],[50,60] ]

for i in range(len(a)):             # 세로 크기(a) 3
    for j in range(len(a[i])) :    # 가로크기 (a[i]) 6
        print(a[i][j], end = ' ')
    print()  # 아무것도 출력하지 않고 줄바꿈
```

```python
# while문 반복문을 한 번 사용해서 2차원 리스트 출력하기 


a = [ [10,20],[30,40],[50,60] ]
i = 0
while i < len(a):  # 반복할 때 리스트의 크기 활용(세로 크기)
    x, y = a[i]   # 리스트의 두 값을 한꺼번에 변수에 할당하는 "언패킹(unpacking)
    print(x,y)    
    i += 1       # 인덱스를 1 증가 시킴 
```

```python
# while 두 번 사용해서 2차원 리스트 출력하기 
a = [ [10,20],[30,40],[50,60] ]

i = 0

while i < len(a):  #가로크기
    j = 0
    while j < len(a[i]):  #세로크기
        print(a[i][j], end = ' ')
        j += 1
    print()
    i += 1
```

```python
# 반복문으로 1차원 리스트 생성하기

a = [] #빈 리스트 생성

for i in range(10):
    a.append(0)  #append로 요소 추가 
print(a)
```

```python
# 반복문으로 2차원 리스트 생성하기 
# 바깥쪽 리스트는 3개, 안쪽 리스트 요소는 2개

a = [] # 빈 리스트 생성

for i in range(3): 
    line = []     # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(2): 
        line.append(0)
    a.append(line)
print(a)
```

```python
# 톱니형 리스트 만들기 - 가로 크기를 알고 있다고 가정.
# 가로크기를 저장한 리스트 a = [ 3,1,3,2,5 ]

a = [ 3,1,3,2,5 ]
b = []

for i in a:     #a(가로 크기)를 저장한 리스트를 반복한다. 총 5번
    line = []  # 안쪽 리스트로 사용할 빈 리스트 생성 
    for j in range(i):
        line.append(0) #  
    b.append(line)
print(b)
    
```

```python
# sorted로 2차원 리스트 정렬하기
# sorted(반복가능한 객체, key = 정렬 함수, reverse = True 또는 False)

# 다음은 학생 정보가 저장된 2차원 리스트를 정렬합니다.
students = [
    [ 'john', 'C', 19 ],
    [ 'maria', 'A', 25 ],
    [ 'andrew', 'B', 7 ]
]
```

```python
# 안쪽 리스트의 인덱스를 1을 기준으로 정렬

# sorted(iterable, key=None, reverse=False)
# sorted(이터러블, 키 = 정렬할 함수, reverse= true는 내림차순 false 오름차순)
# 함수 : key=len, key=str.lower , key= str, upper
# lambda student: student[2]는
# 각 학생 리스트에서 나이만 꺼내서 그걸 기준으로 정렬해
print(sorted(students, key = lambda student : student[1]))
print(sorted(students, key = lambda student : student[2]))
```
