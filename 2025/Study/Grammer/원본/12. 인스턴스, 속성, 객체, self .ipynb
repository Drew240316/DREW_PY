### 1. 인스턴스 정의/역할/메모리 관점 (Instance)


(1) 정의:

  - 클래스라는 설계도를 기반으로 만들어진 “실제 객체”입니다.

  - 예를 들어, Person이라는 클래스를 정의한 후에, p = Person("John", 30)와 같이 호출하면, p는 Person 클     래스의 인스턴스입니다.

(2) 역할:

  - 인스턴스는 그 자체로 하나의 독립된 객체이며, 메모리 상에 할당된 하나의 공간입니다. 

  - 이 공간은 클래스에서 정의한 속성들을 저장하기 위한 컨테이너 역할을 합니다.

(3) 메모리 관점:

  - 인스턴스는 힙(heap) 영역에 생성되며, 각 인스턴스마다 고유한 네임스페이스(예: __dict__)를 가지고 있     습니다. 
  
  - 이 네임스페이스에는 인스턴스의 애트리뷰트 == 속성(데이터, 상태)이 키-값 쌍으로 저장됩니다.

### 2. 속성 정의/역할/메모리 관점 (Attribute)


(1) 정의:

  - 클래스나 인스턴스가 가지는 변수입니다. 이 변수들은 객체의 상태나 데이터를 나타냅니다.

  - 예를 들어, Person 클래스의 인스턴스가 name과 age라는 속성을 가지고 있다면, 이는 해당 객체가 “이
    름”과 “나이”라는 정보를 담고 있다는 뜻입니다.

(2) 역할:

  - 속성은 인스턴스 내부의 데이터를 저장하고, 이 데이터를 통해 객체의 상태를 나타내거나 객체 간의 상호     작용에 사용됩니다.

  - 속성은 클래스 내부에서 정의될 수도 있고, 인스턴스 생성 후에 동적으로 추가될 수도 있습니다.

(3) 메모리 관점:

  - 인스턴스의 네임스페이스(예: p.__dict__)에 속성 이름과 그 값이 저장됩니다.이는 단순한 변수와 같지   
    만, 객체 내부에 소속되어 있는 데이터 항목입니다.

```python
class Person:
    def __init__(self, name, age):
        # 이 두 줄은 인스턴스의 속성(데이터)를 초기화하는 부분입니다.
        # __init__은 객체를 만들 때 **필요한 정보(name, age)**를 바로 넣어주는 함수, 공간 
        self.name = name   # "이 객체의 name 속성에, 입력받은 name 값을 저장해!"
        self.age = age     # "이 객체의 name 속성에, 입력받은 age 값을 저장해!"

    def introduce(self):
        # 메서드는 인스턴스가 가진 속성들을 사용하여 동작(기능)을 수행합니다.
        print(f"안녕하세요, 저는 {self.name}이고, 나이는 {self.age}살입니다.")

# 인스턴스 생성: Person 클래스의 설계도에 따라 실제 객체가 만들어집니다.
p = Person("John", 30)

# 인스턴스 p는 메모리 상에 하나의 객체로 존재하며, 이 객체는 name과 age라는 속성을 가지고 있습니다.
p.introduce()  # 출력: 안녕하세요, 저는 John이고, 나이는 30살입니다.

# 인스턴스 p 내부의 속성 확인
print(p.__dict__)  # 출력 예시: {'name': 'John', 'age': 30}
```

### 8. self는 왜 사용하는가? 

정의) self는 인스턴스 자기 자신을 의미하는데 그 함수를 누가 호출했는지(=그 객체)**를 가리킨다. 

클래스를 호출하여 인스턴스를 생성하면

    p = Person()
    → 이때 p는 Person 클래스의 인스턴스가 됨

    __init__ 메서드가 자동 호출되며,
    첫 번째 인자인 self에 p가 전달됨

    def __init__(self, name):
        self.name = name  # self = p → p.name 이 저장됨
                            인스턴스는 자신의 속성들을 가지게 되고,
                            self를 통해 메서드 안에서도 그 속성들을 자유롭게 사용 가능

                            메서드를 호출할 때도 마찬가지로,
                            해당 인스턴스가 self에 자동으로 전달됨

    p.say_hello()  → def say_hello(self): 에서 self = p

#### 14. (클래스 외부에서 동적으로 속성 추가 가능) init에서 애트리뷰트 없이, 바로 함수에서 애트리뷰트 생성해서 호출하는 방법 
- 인스턴스 생성 > 인스턴스의 다른 메서드 호출 > 인스턴스. 속성 호출 


```python
class Person:
    def greeting(self):
        self.hello = '안녕하세요'  # 메서드 안에서 애트리뷰트를 생성

maria = Person()

# maria.hello 를 이 시점에서 호출하면 에러 발생
# 이유: 아직 greeting() 메서드를 호출하지 않아 hello 속성이 없음
#      즉, 인스턴스는 생성되었지만, __init__에서 hello를 만들지 않았고,
#      메서드를 통해서만 속성이 생성되도록 되어 있음

maria.greeting()  # greeting() 호출 후에야 hello 속성이 maria에게 생김
maria.hello  # 속성은 함수 아님 → () 없이 그냥 maria.hello 로 접근

# 💡 메서드는 maria.greeting() ← () 필수
# 💡 속성은 maria.hello       ← () 없이 그대로
```

파이써닉 한가? 아니오


_init__에서 미리 정의하는 것이 Pythonic 하다.
greeting()에서 속성 만드는 건, 동작 후 부수효과(side-effect)가 생기는 거라 헷갈릴 여지가 있다.

### 15. 특정 속성만 허용하고 싶을 때?

클래스에서 미리 정의된 속성만 허용하고,
그 외의 속성 추가를 막기 위해 사용하는 특별한 속성

- 접근 여부가 아니고, 아예 정의가 되어있느냐 안되어있느냐를 결정
    
    __slots__ = [ '속성1', '속성2' ]  # name, age만 허용(다른 속성은 생성 제한)

```python
class Person:
    __slots__ = ['name', 'age']
    
    def __init__(self, name, age):
        self.name = name 
        self.age = age
        #self.wallet = wallet

    def greeting(self):
        print('hello 반가워!')
    
maria = Person('minji','20')
# maria.wallet 
# AttributeError: 'Person' object has no attribute 'wallet'
# "접근이 안 된다" → 정확히는 '그 속성이 아예 없다'는 뜻
```

##### (설명) slot을 왜 쓰는데?

 - 위 코드는 내부적으로 p.__dict__에 이렇게 저장. 즉, 속성을 추가하면 파이썬은 __dict__에 키-값으로   저장함 그래서 아무 속성이나 막 추가할 수 있음.


 - 근데 __slots__을 선언하면?

  이 클래스는 이제 __dict__ 자체가 없음! 그래서 name과 age만 메모리 공간에 고정 슬롯으로 할당되고,
  그 외 속성은 에러가 나면서 못 넣음

  - 결국, 이럴 때 사용.!

    가. 메모리 절약	__dict__ 사용 안 해서, 수천 개 인스턴스일 때 효과 큼 
    나. 속성 제한	불필요한 속성 추가 방지 → 코드 안전성 증가  
    다. 성능 향상	__dict__보다 빠르게 속성 접근 가능 (조금 더 빠름)

  -  종합 : 대량의 객체 생성 시, 속성 구조 고정할 때, 성능 최적화할 때

### 16. 비공개 속성(private attribute) 사용해서 클래스 안에서만 접근하기
###### 비공개 속성 == 고정된 값

  - 속성명 앞에 **__(언더스코어 두 개)**를 붙이면, 외부에서 직접 접근이 막히고, 클래스 내부에서만 사
    용 가능하도록 "캡슐화" 됨, 


  - 왜 사용하는가? (사용 목적)

    가. 데이터 보호	실수로 외부에서 값이 변경되지 않도록 막음
    
    나. 캡슐화	      클래스 내부 구현을 숨기고, 외부에서는 오직 “제공된 메서드”로만 접근하게 강제
    
    다. 유지보수성 향상	클래스 내부 로직이 변경되어도 외부 코드는 그대로 둘 수 있음
    
    라. 의도 명확화	"이 속성은 외부에서 만지지 마세요!" 라는 신호를 줌
    
  - 주된 에러
    
    **클래스 바깥** 에서 비공개 속성에 접근하면 에러 발생
    
    AttributeError : ~~~ '__속성명'
  
  - 속성뿐아니라, 메서드도 비공개 속성 사용 가능

```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet ## 변수 앞에 __를 붙여서 비공개 속성 만듬 
    
    def pay(self, amount):
        self.__wallet -= amount 
        print('이제 {0}원 남았네요.'.format(self.__wallet))

maria = Person('마리아', 20,'서울시 영등포구',10000)
maria.pay(3000)
```
