```python
# 람다 표현식 == 익명 함수 (def로 이름 붙이지 않고 바로 사용 가능)
# lambda 매개변수 : 표현하는 식 , 굳이 이름을 안붙이는 일회성 함수로 많이 사용! 
#  map , filter, sorted 함수랑 같이 자주 사용

# 사용자 정의함수에서는 매개변수와 인수를 사용해서 값을 받음 
# 매개변수 : 호출을 정의할 때 넘겨 받는 값
# 인수 : 호출을 넘겨줄 때 넘겨주는 값.
```

```python
# 람다 표현식으로 함수 만들기
# 람다 표현식 만들기 전에 숫자를 받은 뒤 10을 더해서 반환하는 함수를 만들어본다.
def plus_ten(x):
    return x + 10

plus_ten(1)
```

```python
# lambda 매개변수 : 표현식

plus_ten = lambda x : x + 10  # 매개변수 x 하나를 받고, x에 10을 더해서 반환한다는 뜻입니다.
plus_ten
```

```python
# lambda 표현식 자체를 호출하기 
# lambda 표현식은 변수를 할당하지 않고 괄호와 인수를 사용하여 바로 호출 가능하다.
# (lambda 매개변수 : 표현식)(인수)
(lambda x : x+10)(1)
```

```python
# 람다 표현식 안에서 변수를 선언할 수 없다.! 
# (lambda x : y = 10; x+y)(1) 
## SyntaxError: invalid syntax  #오류 (람다 안에 표현식을 사용 가능하고, 변수y=10)는 못 씀)
# y = 10; (lambda x: x+y)(1)  #정상 (람다 표현식 밖에 있는 변수는 사용 가능)
y = 10
(lambda x: x+y)(1)
```

```python
# 람다 표현식을 인수로 사용하기
# 람다표현식을 사용하는 이유는 함수의 인수 부분에서 간단하게 함수를 만들기 위함이다. 

def plus_ten(x):
    return x + 10

list(map(plus_ten, [1,2,3]))  
```

```python
# 파이썬의 함수 인수
# 인수는 호출할 때 넘겨주는 값으로, 함수의 인수에 함수가 들어갈 수 있다.
# 물 -> 주전자 -> 컵 
#함수와 인수 구조:

# map()의 인수(argument):

# plus_ten → 함수

# [1, 2, 3] → 리스트

# list()의 인수:

# map(plus_ten, [1, 2, 3])의 결과 (즉, map 객체)
```

```python
# 람다 표현식으로 변경
# map(함수, 반복 가능한 값) 
## 반복가능한 값 = 리스트, 튜플, [문자_서식_포맷_정규], range , dict, set
list(map(lambda x:x+10, [1,2,3]))
```

```python
# 람다 표현식 풀어보기 
def plus_ten(x):
    return x + 10

result = list(map(plus_ten, [1, 2, 3]))
print(result)  # [11, 12, 13]
```

```python
# 람다 표현식과 map filter reduce 함수 활용하기
## 람다 표현식에 조건부 표현식 사용하기 


## 조건부 표현식 스타일 1 (if-else)
### lambda 매개변수들 : 식1 if 조건식 else 식2 , 이터러블 객체


## 조건부 표현식 스타일 2 (if - elif - else)
#### lambda x: 결과1 if 조건1 else 결과2 if 조건2 else 결과3
#### 조건부 표현식 스타일 2는 길어지면 보기 불편하므로 if elif else를 사용자 정의함수를 통해 사용 권장
#### 람다 조건부 표현식은 elif 사용 안함 
#조건1이 참이면 → 값1
#조건1이 거짓이고 조건2가 참이면 → 값2
#조건1, 조건2 모두 거짓이면 → 값3


# 같이 사용하는 함수들?
## map, filter, reduce
## map - 리스트 등의 반복 가능한 객체를 여러개 넣을 수 있고, 매개변수를 2개 지정한다.
#      - 그리고 map에 람다 표현식을 넣고 그 다음에 리스트(반복가능한 객체) 두 개를 콤마로 
#        구분해서 넣는다.
# map(함수, 반복가능한객체)
# filter(함수, 반복가능한객체) : 반복 가능한 객체에서 **특정 요건**에 맞는 요소만 가져온다. 
#                               **참 요소**만 가져오고 거짓된 요소 버리기
                        # 항상 list(filter(,)) 형태로 써야 출력된 값이 보임. 
# reduce (함수, 반복가능한객체) : 
```

```python
# 조건부 표현식 스타일 2를 사용자 정의 함수로 풀어봄
def 평가(x):
    if x > 90:
        return "A"
    elif x > 80:
        return "B"
    else:
        return "C"
```

```python
# map에 객체를 여러 개 넣기
a = [1,2,3,4,5]
b = [2,4,6,8,10]
list(map(lambda x, y : x*y, a, b))
```

```python
# filter 사용하기 

a = [8,3,2,10,15,7,1,9,0,11]
list(filter(lambda x: x>5 and x < 10, a))
```

```python
# filter 함수를 def로 풀어보기 

def f(x):
    return x > 5 and x < 10

a = [8,3,2,10,15,7,1,9,0,11]
list(filter(f,a))   # 함수 호출 방법과 다르다. 일반적인 함수 호출 f(전달하는 인자)
```

```python
# reduce 사용하기
# 빌드인 함수 아님
# 객체의 각 요소를 지정된 함수로 처리한 뒤, 이전 결과와 누적해서 반환하는 함수
from functools import reduce
#reduce(함수, 반복가능한객체)

def f(x,y):
    return x+y

a = [1,2,3,4,5]
from functools import reduce
reduce(f,a)
```
