# 함수에서 위치인수와 키워드 인수, 리스트, 딕셔너리 언패킹 방법 (편한 기능)

```python
## 위치 인수와 리스트 언패킹 사용하기

print(10,20,30)   # print에서  10,20,30 순으로 넣었으므로 출력될 때도 10,20,30으로 출력됩니다. 
```

```python
### 위치 인수를 사용하는 함수를 만들고 호출하기 (숫자를 넣자)

def print_numbers(a,b,c):
    print(a)
    print(b)
    print(c)
    
print_numbers(10,20,30)
```

```python
### 언패킹 사용하기 (리스트나 튜플을 넣자)
# 함수(*리스트)
# 함수(*튜플)
### 리스트나 튜플 앞에 *을 넣어서 언패킹=포장을 풀어주세요
# * 연산자는 리스트나 튜플의 요소를 하나씩 **위치 인자(positional arguments)**로 함수에 전달할 때 사용합니다.
### 리스트 변수나 리스트 요소 개수를 넣어야 됩니다. 단,  매개변수와 리스트의 요소 개수는 같아야됩니다.
### 주의
#### 고정 인수와 가변 인수를 함께 사용할 때는 고정 매개변수를 먼저 지정하고, 그다음 매개변수에 *붙어야함

i = [10,20,30]
print_numbers(*i)
```

```python
print_numbers(*[10,2,30])
```

```python
# 가변 인수 함수 만들기 =  *argumets(args, 인수)
# 위치 인수와 리스트 언패킹은 어디에 사용할까요?
# 이 기능들은 인수의 개수가 정해지지 않은 가변인수(variable argument)에 사용합니다.
# 즉, 같은 함수에 인수 한 개를 넣을 수도 있고, 없을 수도 있고, 열 개를 넣을 수도 있습니다.
# def 함수이름(*매개변수):
    #코드
```

```python
def print_numbers(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers(10)
```

```python
print_numbers(10,20,30,40)
```

```python
x = [10]
print_numbers(*x)
```

```python
y = [10,20,30,40]
print_numbers(*y)
```

```python
#위치 인수와 가변인수를 파라미터에 넣고 호출하기
def print_numbers(a,*args):
    print(a)
    print(args)

print_numbers(1,10,30,40,50)
```

```python
print_numbers(1,10,20)
```

```python
print_numbers(*[10,20,30])
```

```python
## 키워드 인수 사용하기
### 함수의 사용 방법을 익힐 때 인수의 순서와 용도를 함께 외웁시다. 
### 하지만 각각의 인수가 무슨 용도인지 알기 어려울 때는 "키워드 인수"를 사용합니다.


# 키워드 인수 x 

def personal_info(name, age, address):
    print('이름: ', name),
    print('나이: ', age),
    print('주소: ', address)
    
personal_info('홍길동',30,'서울시 용산구 이촌동')
```

```python
## 키워드 인수 o
### 순서가 틀려도 상관 없음.!

personal_info(age=40,  name = 'minji', address = '서울시 용산구')
```

```python
print(10,20,30, sep= ' ', end = ' ')  #sep end도 키워드 인수입니다. 
```

```python
## 키워드 인수와 딕셔너리 언패킹 사용하기
### 딕셔너리 앞에 에스터리스크 두개를 붙여서 함수에 넣어줍니다.
### 함수(***딕셔너리)
### 딕셔너리 변수 대신 딕셔너리 앞에 바로 ** 붙여도 언패킹 가능
### 딕셔너리 이름과 개수가 다르면 호출 불가 
def personal_info(name, age, address):
    print('이름:', name)
    print('나이:', age)
    print('주소:', address)
```

```python
x = { 'name' : '홍길동', 'age' : 30, 'address' : '서울시 용산구 이촌동' }
personal_info(**x)
```

```python
personal_info(**{ 'name' : '홍길동', 'age' : 30, 'address' : '서울시 용산구 이촌동' })
```

```python
personal_info(**{ 'last_name' : '홍길동', 'age' : 30, 'address' : '서울시 용산구 이촌동' })

## personal_info() got an unexpected keyword argument 'last_name'
## 이름이 다르다!
```

```python
personal_info(**{ 'age' : 30, 'address' : '서울시 용산구 이촌동' })

##: personal_info() missing 1 required positional argument: 'name'
## 이름이 없다.
```

```python
## **을 두 번 사용하는 이유? 딕셔너리는 키 - 값 쌍 형태로 지정된다.
## * 이 1개면 키가 출력
## **이 두 개면 키 값 두번 언패킹한다는 뜻
```

```python
personal_info(*{ 'name' : '홍길동', 'age' : 30, 'address' : '서울시 용산구 이촌동' })
```

```python
personal_info(**{ 'name' : '홍길동', 'age' : 30, 'address' : '서울시 용산구 이촌동' })
```

```python
# 전달한 호출 인자의 키값이 있는지 확인하고 출력 (가변인수, in)
def personal_info(**kwargs):
    if 'name' in kwargs:     # in으로 딕셔너리 안에 특정 키가 있는지 확인          
        print('이름: ', kwargs['name']) 
    if 'age' in kwargs:
        print('나이: ', kwargs['age'])
    if 'address' in kwargs:
        print('주소: ', kwargs['address'])
```

```python
# 고정 인수와 키워드 인수를 함께 사용하기

def personal_info(name, **kwargs):
    print(name)
    print(kwargs)

personal_info('홍길동')
```

```python
personal_info('홍길동',age = 30, address = '서울시 용산구 이촌동')
```

```python
personal_info(**{'name':'홍길동','age':30,'address':'서울시 용산구 이촌동'})
```

```python
# 위치 인수와 키워드 인수를 함께 사용하기.
# 위치 인수는 1 2 3 튜플로 받고
# 키워드 인수는 sep = ':', end = ''으로 받기

#주의 ) 위치 인수>키워드 인수 순서로 작성해야함.
#사례 ) print(10,20,30, sep= ' ', end = ' ')


def custom_print(*args, **kwargs):   
    print(*args, **kwargs)

custom_print(1,2,3, sep=':', end='')   # 위치 인수, 키워드 인수 같이 사용함.


# 만약에, 고정인수, 위치인수, 키워드 인수를 사용하려면?
# def custom_print(a,b, *args, **kwargs) 처럼 고정 매개변수 *args **kwargs순서로 해야함/
```

```python
# 매개변수에 초깃값 지정하기( 파라미터 = '비공개'로 해서, 호출 시 고정 인수를 생략하고 호출 가능 )
# 왜? 주로 사용하는 값이 있으면서 가끔 다른 값을 사용해야 할 때 활용
# 예를들어 print함수의 sep는 초깃값이 ' '공백으로 지정되어 있어서 대부분 그대로 사용하고 가끔 sep에 
# 다른 값을 넣어서 사용합니다. 


def personal_info(name, age, address = '비공개') : #초깃값 지정
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

    
```

```python
# address는 초깃값이 있으므로, personal_info는 address 부분을 비워두고 호출 가능
personal_info('홍길동', 30)
```

```python
# 매개변수에 초깃값이 지정되어 있더라도 값을 넣으면 해당 값이 전달 됩니다.
personal_info('홍길동', '서울특별시 광진구')
```

```python
# 초깃값이 지정된 매개변수의 위치 

def personal_info(name,  address = '비공개', age) : #초깃값 지정
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)'

    # non-default argument follows default argument
    # 초기값이 지정된 매개변수는 항상 뒤쪽에 몰아 넣어주어야 한다.
    # 왜냐? personal_info('홍길동',30)을 호출했을 때  30이 어디로 들어가야하는지 알 수 없다.
    #      address에 들어가려니 age 부분이 비어버린다. 
```
