```python
# 8. continue를 사용해서 입력한 숫자 까지 홀수 출력하기

mimi = int(input('입력:'))

for i in range(mimi+1):
    if i % 2 == 0:
        continue
    print(i)
    

    
```

```python
# 9. break를 사용해서 입력한 횟수대로 반복하기.  

for i in range(30):
    print(i)
    if i == 10:
        break
    i += 1
```

```python
# 10. 3의 배수는 제외하고 1~10까지 출력해라.

for i in range(1,11):
    if i % 3 ==0:
        continue
    print(i)
```

```python
# 11. 무한루프를 사용하여 출력하되 값이 하나씩 감소하고, 10을 만나면 멈춰라. 

i = 30

while True:
    print(i)
    if i == 10:
        break
    i-=1
```

```python
# 12. 무한 루프 시 3으로 끝나는 숫자를 출력하고, 30이 넘으면 탈출해라.

i = 0

while 1 :
    i += 1
    if i % 10 != 3:
        continue
    if i > 30:
        break
    print(i)
```

```python
# 13. 사각형으로 별 출력하기
#     4*5 사각형 형태로 별 그리기 

for i in range(4):
    for j in range(5):
        print('*', end = ' ')
    print()

```

```python
# 14. 계단식으로 별 출력하기
#     5칸의 계단을 만들고 싶습니다. 

for i in range(5):
    for j in range(5):
        if j <= i :
            print('☆', end = ' ')
    print()
```

```python
# 15. 거꾸로된 계단식으로 별 출력하기
#     1층이 별이 1개가 있습니다.2층은 별 2개..

for i in range(5):
    for j in range(5):
        if i <= j:
            print('☆', end = ' ')
    print()

    
# range는 항상 0부터 시작함을 꼭 생각하자. 
```
