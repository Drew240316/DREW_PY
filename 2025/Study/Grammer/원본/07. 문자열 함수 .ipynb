# 서식 지정자

```python
# [문자_서식_포맷_정규] 서식 지정자와 포매팅 사용하기
# 파이썬은 다양한 방법으로 문자열을 만들 수 있다.
# 그중에서 서식 지정자(format specifier)로 문자열을 만드는 방법과
#         format 메서드로 문자열을 포매팅(string formating)에 대해 알아보겠습니다.
```

```python
# 서식 지정자로 [문자_서식_포맷_정규] 넣기 또는 문자열을 바로 지정하지 않고 변수를 지정하기
# '%S' % '[문자_서식_포맷_정규]'

'I am %s.' % 'james'
```

```python
name = 'maria'

'I am %s.' % name
```

```python
# 서식 지정자로 숫자 넣기 
# '%d.' % 숫자

'I am %d years old' % 20 
```

```python
# 서식 지정자로 소수점 표현하기 
# '%f' % 숫자 

'%f' % 2.3
```

```python
# '%.자릿수f' % 숫자 

'%.4f' % 2.3
```

```python
# 서식 지정자로 [문자_서식_포맷_정규] 정렬하기
# %길이s
# 서식 지정자와 숫자를 조합하여 문자열을 정렬하는 방법
# %뒤에 숫자를 붙이면 문자열을 지정된 길이로 만든 뒤 오른쪽으로 정렬하고 남는 공간을 공백으로 채움
# [문자_서식_포맷_정규] 오른쪽 정렬은 자릿수가 다른 숫자를 출력할 때 유용

'%10s' % 'python'
```

```python
for i in [1, 10, 100, 1000]:
    print('%5d' % i)  # 5칸 확보 후 오른쪽 정렬
    
    
# 데이터를 비교하거나 테이블을 쉽게 보기 위해 주로 사용 
```

```python
# 왼쪽 정렬
# %-길이s

'%-10s' % 'python'
```

```python
# 서식 지정자로 [문자_서식_포맷_정규] 안에 값 여러 개 넣기
# '%d%s'%(숫자, '[문자_서식_포맷_정규]')

'Today is %d %s. ' % (3, 'April') 
```

# format 함수

###### {} 대괄호 안을 포맷팅 이라 함.


```python
# format 메서드로 값을 여러개 넣기

'Hello, {0} {2} {1}'.format('python', 'script', 3.6)
```

```python
# format 메서드에서 인덱스 생략하기

'Heloo, {} {} {}'.format('Python','Script', 3.6)
```

```python
# format 메서드에서 인덱스 대신 이름 지정하기

'Hello. {language} {version} '.format(language='Python', version = 3.6)
```

```python
# [문자_서식_포맷_정규] 포매팅에 변수를 그대로 사용하기
language = 'python'
version  = 3.6

f'Hello World, {language} {version}'
```

```python
# format 메서드로 [문자_서식_포맷_정규] 정렬하기
# [문자_서식_포맷_정규] 서식 지정자나 ljust, rjust, center로 정렬같이 [문자_서식_포맷_정규] 포맷팅으로도 가능하다.
# '{인덱스:<길이}'.format(값)
# 부등호가 왼쪽으로 가리키면 왼쪽, 오른쪽으로 가리키면 오른쪽 

'{0:<10}'.format('python')
```

```python

# 정수와 실수 앞에 0을 넣어서 숫자 개수를 맞추는 방법이 있다
# %와 d 사이에 0과 숫자 개수를 넣어주면 자릿수에 맞춰서 앞에 0이 들어갑니다.
# %03d로 지정하면 1은 '001', 35는 '035'가 됩니다. {}를 사용할 때는 인덱스나 이름뒤에 콜론을 붙이고 지정 가능합니다.

# '%0개수d' % 숫자
# '{인덱스:0개수d}'.format(숫자)

'%03d' % 1
```

```python
'{0:03d}'.format(35)   
```

```python
'%08.2f' % 3.6
```

```python
'{0:08.2f}'.format(150.37) 
```

```python
# '{인덱스:채우기정렬길이}'.format(문자열에 넣을 값)

'{0:0<10}'.format(15)


# 0 → 인덱스 0번째 값 (즉, format(15)의 15를 가리킴)

# : → 포맷팅 시작

# 0 → 빈 자리를 0으로 채워라

# < → 왼쪽 정렬하라

# 10 → 전체 길이를 10으로 맞춰라
```

```python
'{0:0>10}'.format(15)
```

# 금액에서 천단위로 콤마 넣기 

```python
# 금액에서 천다위로 콤마 넣기
# format(숫자,',')

format(1493500, ',')
```

```python
# format 함수는 서식지정자와 함께 사용 가능
# 콤마를 넣은 숫자를 오른쪽으로 정렬한다. 길이는 20

'%20s' % format(1439500, ',')
```

```python
# 포매팅에서 콤마를 넣으려면 다음과 같이 :(콜론) 뒤에 ,(콤마)를 넣는다
'{0:,}'.format(143950)
```

```python
# 만약 정렬을 하고 싶다면, 정렬 방향과 같이 길이 뒤에 콤마를 지정한다.
'{0:20,}'.format(1439500)
```
