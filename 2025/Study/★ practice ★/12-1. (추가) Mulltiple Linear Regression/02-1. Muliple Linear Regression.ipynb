#### 원 핫 인코딩


```python
import pandas as pd

dataset = pd.read_csv("C://Users//wkdal//Desktop//archive//PythonMLWorkspace(LightWeight)//ScikitLearn//MultipleLinearRegressionData.csv")
```

```python
dataset.head()
```

```python
import pandas as pd

X= dataset.iloc[:,:-1].values #전체 로우 중에서 컬럼은 처음부터 끝 직전까지
Y = dataset.iloc[:,-1].values # 전체 로우 중에서 컬럼은 맨 마지막 것
```

```python
X

# 컬럼은 공부 시간 , 결석 횟수, 공부 장소
```

```python
# 원하는 것: 다중 공산성 문제를 해결하기 위해, 세 개중에서 1개만 삭제한 두 개 컬럼만 추가한다.
```

```python
from sklearn.compose import ColumnTransformer # 컬럼 트랜스포머 클래스
#서로 다른 전처리 작업을 열(column)별로 따로 적용할 수 있도록 도와주는 유틸리티야.
# 특히 숫자열/범주형이 섞인 데이터 전처리
from sklearn.preprocessing import OneHotEncoder # 원핫 인코딩 적용에 필요한 클래스
```

```python
# 컬럼 트랜스포머 클래스로 '컬럼 트랜스포머 객체 만들기

ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(drop='first'),[2])],
                       remainder='passthrough')
# transformers : 전처리기(preprocessors)" 리스트야. 즉, 데이터의 특정 열에 어떤 변환(transform)을 적용할지를 정의하고, 튜플 형태()로 적는다.
    # 'encoder' : 이 변환 작업의 이름, 즉 어떤 변경을 가할지 명시
    # OneHotEncoder :적용할 전처리기 (여기선 범주형을 원-핫 인코딩하고 다중 공산성 문제를 없애기 위해 드롭해서 첫 항은 생략 ex.4개 피처면 3개만 쓰고, 10개면 9개만 쓰고..이런식..으로 다중공산성 문제 해결)
    # drop='first'
    # 어떤 데이터를 가지고 원핫인코딩을 적용할지 컬럼 인덱스를 정한다. [2]는 2번째인덱스
    # remainder 원핫 인코딩을 적용하지 않는 나머지 컬럼을 어떻게 적용할것이냐는 말에 pass를 써서
      #그대로 둔다는 말.


# x값을 적용 -> 즉, 원핫 인코딩을 통해 컬럼이 1 0 0 1이런식으로 바뀜
X = ct.fit_transform(X)
```

```python

X
```

```python
# 실행 결과
# 맨 앞쪽에 컬럼 2개가 추가됨을 알 수 있다.(공부장소빼고 나머지 2개 추가됨)
# 1 0 HOME
# 0 1 LIBRARY
# 0 0 CAFE
```

##### 데이터 세트 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2
                                                    ,random_state = 0)
```

#### 학습 (다중 선형 회귀)

```python
# 모델 만들기!
from sklearn.linear_model import LinearRegression #리니어리그레션 클래스 가져오고
reg = LinearRegression() # 리니어 리그레션 객체 만들고
reg.fit(X_train, Y_train) #앞에서 만들었던 훈련 데이터로 바로 학습
```

#### 예측값 실제값 비교(테스트세트)

```python
# 모델 학습 (x_train y train) : 훈련 데이터로 모델을 학습시킨다.
# 예측 ( y pred ) : 학습된 모델로 테스트 입력에 대한 예측 수행
# 평가 ( y test, y pred ) :	예측값(y_pred)과 실제값(y_test) 비교하여 성능 측정

y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
Y_test


# 예측값(Y_PRED)와 실제값(Y_TEST)를 비교해보니,
# 실제값은 90점인데, 모델은 92점으로 예측했고..... 38점은 근사하게 예측해서 모델이 괜찮음.
```

```python
# 기울기 (각 변수의 영향력 (기울기))
reg.coef_
# 독립변수 4개에 대해서 나오고 있고, 이것의 의미는
# 집에서 공부하면 -5.8 도서관은 -1. 카페에서 둘다 0이니깐 그대로 됨.
# 카페에서 공부할 때가 가장 도움이 됨.
# 공부시간은 1시간씩 증가할 때마다 점수가 10점씩 오름
# 결석은 한 번 할 때마다 약 1.64점씩 안좋은 영향을 줌.
```
