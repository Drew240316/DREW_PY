### 공부 시간에 따른 시험 점수

```python
# 데이터 시각화
import matplotlib.pyplot as plt
# 데이터 가공
import pandas as pd
```

```python
# 데이터 읽어오기
Study_sample = pd.read_csv("C://Users//wkdal//Desktop//archive//PythonMLWorkspace(LightWeight)//ScikitLearn//LinearRegressionData.csv")
```

```python
Study_sample.head()
```

```python
# 보통 작업할 때 독립변수 x가 굉장히 많을 예정이고 종속변수는 하나이다.
# 동일한 템플릿으로 작업하도록 아래와 같이 작성
# X = dataset.iloc[row, col].values
    # row에 콜럼을 사용해서 모든 row를 가져와라.
    # col 부분은 row를 제외한 한가지, 여기선 score를 제외한 한가지를 가져올 것이라 -1을한다.
      # 전체부터 마지막 직전까지 값을 가져오는 것이다. (모든 행에서 마지막 열 제외)
    # vlaes는 pandas의 DataFrame 또는 Series에서 NumPy 배열(ndarray) 형태로 데이터를 추출하 #는 속성
```

```python
X = Study_sample.iloc[:, :-1].values
```

```python
# col 모든 행에서 마지막 열 하나만 선택
Y = Study_sample.iloc[:,-1].values
```

```python
X, Y
```

```python
# 선형 회귀 모델 만들기 (불러오고, 모델 객체 만들기)
from sklearn.linear_model import LinearRegression
    # 에스케이런 점 리니어 언더바 모델 임포트 대문자 엘 리니어 대문자 알 리그레이션
reg = LinearRegression()
    # reg에는 리니어 레그레이션 모델 객체 생성함.
# 모델에는 fit이 있다. fit을 통해 학습을 하면서 모델을 생성한다. (학습- 모델 생성)
reg.fit(X,Y)
```

```python
# 여기에 어떤 데이터값(X)을 넣어서 예측값(Y)를 보고,
# X값? array([[ 0.5],
     #   [ 1.2],
      #  [ 1.8], .....

# Y값? array([ 10,   8,  14,  26,  22,  30,  42,  48,  38,  58,  60,  72,  62,
       #  68,  72,  58,  76,  86,  90, 100], dtype=int64))
```

```python
# 추정되는 예측점수(Y)는 뭘까?
    # reg모델에는 predecit()라는 함수가 있어서, X를 넣어서 예측값을 만든다

y_pred = reg.predict(X) # x에 대한 예측값
y_pred
```

```python
# 결과
    #  5.00336377를 보면, x가 0.5시간 공부했을 때, 예측점수는 5.000점이고..
    # 1.2시간 공부했을 때 12점...나오는 것을 예측한 것.
```

```python
# 눈으로 보기 편하기 위해 데이터 시각화를 해보자.
plt.scatter(X,Y,color= 'blue') #X와 Y그래프를 산점도로 하고, 색깔은 파랑색
plt.plot(X,y_pred, color = 'green') # 플롯은 선 그래프라는 뜻, X값과 Y값은 X를 넣었을 때 예측값이라 y_pred, 초록색으로 설정
plt.title('Score by hours') # 제목은 시간에 따른 점수.
plt.xlabel('hours') #xlabel은 x축이란 뜻
plt.ylabel('score') #ylabel은 y축 이름 이란 뜻
plt.show() # 시각화를 보여줘!
```

```python
# 결과
# 공부를 함에 따라 점수를 확인할 수 있고
# 리니어 리그레이션 모델을 통해 가장 최적의 선 한개를 볼 수 있다.
# 그러면 9시간 공부하면 예상되는 점수는? 아래에 출력해보자
```

3. 리니어리그레이션 객체를 만들어서 핏 함수를 통해 모델 생성하고 학습을 시킴(모델 만들어짐)
4. 그 모델에서 프레딕트 함수 호출시켜서 x들어가는 값에 대해 예측 결과가 나온다.
5. (4번은 모든 x값을 넣은 것이고) 각 한개씩 보고 싶으면 이차원배열로 만들어서 본다
6. 그 다음, 산점도, 선그래프를 통해 각 점들과 최적의 선을 표현했다.

## 경사하강법 (gradient descent)

```python
from sklearn.linear_model import SGDRegressor # SGD : Stochastic Gradient Descent 확률적 경사하강법
```

```python
# 모델 객체 sr
sr = SGDRegressor(max_iter=100, eta0=0.0001, random_state=0,verbose=1)
# max_iter (맥스 이터) : 훈련세트 반복 횟수, 에포크 횟수를 1000 정도
    # 에포크 횟수에 다다랐는데, 반복횟수 값을 키워보는게 어떤지 경고도 나옴
       # ConvergenceWarning: Maximum number of iteration reached before convergence. Consider increasing max_iter to improve the fit.

# eta0(이타제로) : 학습률은 0.001로 디폴드 값, 1e-3으로 지수 표기법으로 할 수 있다.
    #1e-3 : 0.001 (10^-3) 10의 마이너스 3승
    #1e-4 : 0.0001 (10^-4) 10의 마이너스 4승
    #1e+3 : 1000 (10^3)
    #1e+4 : 10000 (10^4)
        # 에포크의 수를 줄이고 늘릴 수 있으며, loss가 떨어지는 것을 확인할 수 있다.
# random_state =0
# vervose = 1로해서 훈련세트를 반복하면, 어떻게 되는지 보여줌.

sr.fit(X_train, Y_train)
```

```python
# 눈으로 보기 편하기 위해 데이터 시각화를 해보자.
# 80대 20대부분에서 20만.(검증하기 위한 부분)
plt.scatter(X_train,Y_train,color= 'blue') #X와 Y그래프를 산점도로 하고, 색깔은 파랑색
plt.plot(X_train,sr.predict(X_train), color = 'green') # 플롯은 선 그래프라는 뜻, X값과 Y값은 X를 넣었을 때 예측값이라 y_pred, 초록색으로 설정
plt.title('Score by hours(test_data,SGD)') # 제목은 시간에 따른 점수.
plt.xlabel('hours') #xlabel은 x축이란 뜻
plt.ylabel('score') #ylabel은 y축 이름 이란 뜻
plt.show() # 시각화를 보여줘!
```

```python
# 기울기 y절편을 확인해라.(훈련 데이터의 평가 점수와 비교한다.)
sr.coef_, sr.intercept_
```

```python
# test세트를 통한 확률적 경사하강법 모델 평가

sr.score(X_test, Y_test)

```

```python
sr.score(X_train, Y_train)

```
