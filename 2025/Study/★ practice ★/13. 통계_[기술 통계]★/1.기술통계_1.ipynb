### 1.1 정형화된 데이터의 요소

#### 주요 개념 (종, 수범, 처)
    - 일반적으로 소프트웨어에서는 데이터를 종류별로 구분한다.
    - 데이터 유형에는 수치형(연속,이산)과 범주형(이진,순서)이 있다.
    - 소프트웨어에서는 데이터 종류를 정하는 것은 해당 데이터를 어떻게 처리할지를 정하는 것과 같다.
    
#### 용어 정리 ( 정, 수, 범, 이, 순 )
    1. 정형 데이터 : 수치형, 범주형 데이터
    2. 수치형 데이터 : 풍속이나 지속시간은 '연속형'데이터, '사건의 발생 빈도', '횟수'와 같은 
                      '이산 데이터'가 있다.
    3. 범주형 데이터 : TV 스크린 종류 (플라즈마, LCD, LED)나 도시명(대전, 부산, 서울)과 같이
                       범위가 정해진 값들을 갖는 경우. (이진 데이터, 순서형 데이터 포함)
    4. 이진 데이터 : 이진 데이터는 이러한 '범주형 데이터'중에서도 0과 1, 예/아니오, 
                     참/거짓과 같이 둘 중 하나를 갖는 경우
    5. 순서형 데이터 : 범주 안의 '값'들이 '순위'를 갖는 범주형 데이터, 평점(1,2,3,4,5)가 해당.

#### 범주형인지 확실히 구분하면 따라오는 3가지 이점 ( 차모, 인, 소 )
    1. 범주형 데이터라는 정보는 소프트웨어가 '차트 생성'이나 '모델 피팅' 등 '통계 분석'을 수행하는 방식을 결정.
       sklearn.processing.OrdinalEncoder로 순서형 데이터를 지원
    2. 관계형 데이터베이스에서처럼 저장소와 인덱싱을 최적화하는 데 사용한다. (자료구조 = 저장, 알고리즘 = 탐색)
    3. 범주형 변수가 취할 수 있는 값들은 소프트웨어적으로 처리 가능하다. 
    

### 1.2 테이블 데이터

#### 주요 개념 ( 레, 피 )
    1. 데이터 과학의 기본이 되는 데이터 구조는 행과 열이 각각 레코드 변수(피처)를 의미한다.
        - 피처(예측 변수) = 타겟(종속 변수)
        - 데이터 과학 : 각 행 = 샘플 , 통계학 : 여러 행 = 샘플
        
#### 용어 정리 ( 데, 피, 결, 레)
    1. 데이터 프레임 : 통계와 머신러닝 모델에서 가장 기본이 되는 테이블 형태의 데이터 구조
    2. 피처 : 일반적으로 테이블의 각 열이 하나의 피처 (유의어 : 특징, 속성, 입력, 예측변수, 변수)
    3. 결과 : 데이터 과학은 어떤 결과를 예측한다. 이처럼 결과를 예측하기 위해 피처를 사용한다.
    4. 레코드 : 일반적으로 테이블의 각 행은 하나의 레코드를 의미한다. 
                (유의어 : 기록값, 사건, 사례, 예제, 관측값, 패턴, 샘플)
    

#### 1.3 위치 추정 
  " 수천, 수만가지 데이터 변수들의 대표가 어디쯤 위치하나? = 대표값은? = 추정값은? "
    
#### 주요 개념 (평, 중절)
    1. 가장 기본적인 위치 추정 기법은 평균이다. 하지만 극단값(특잇값)에 민감할 수 있다.
    2. 중간값, 절사평균과 같은 다른 방법들이 특잇값이나 이상 데이터에 덜 민감하므로 좀 더 로버스트하다. 

#### 용어 정리 (평, 가, 중, 가, 절, 로, 특, 참 , 추 측)
    1. 평균(mean) : 모든 값의 총합을 개수로 나눈 값(유의어 : 평균average)
    2. 가중평균 : 각 데이터 값 xi에 사용자가 지정한 가중치 wi를 곱한 값들의 총합을 다시 가중치의 총합으로 나눔.
    3. 중간값 : 데이터에서 가장 가운데 위치한 값(유의어 : 50번째 백분위 수)
    4. 가중 중간값 : 데이터를 정렬한 후, 누적 가중치의 합이 전체 가중치의 50%를 초과하는 첫 번째 값
    5. 절사평균 : 값들의 크기 순으로 정렬한 후, 양끝에서 일정 개수의 값들을 삭제한 뒤 남은 값들.
    6. 로버스트하다 : 극단값들에 민감하지 않다는 것을 의미(유의어 : 저항성 있다.)
    7. 특잇값 : 대부분의 값과 매우 다른 데이터 값(유의어 : 극단값) 
    8. 참값 (True value, 이론값) : 우리가 알고 있는 진짜 상태 (대한민국 남성 전체의 평균 키)
    9. 추정값 (estimate) : 그 참값을 직접알 수 없어서, 일부 데이터를 모아서 계산으로 추정하는 값
      이런 추정값들이 정확도, 정밀도, 재현율, 평균 등으로 표현될 때 측정 지표라한다.
      결국 추정값과 측정지표는 같은 말이다. 
   

##### 1.3.1 평균 ( 평, 절, 가 , 가)
  " 대표성을 얻는 방법과 극단값의 관계는? 가중 평균을 왜 사용하지? "
###### 평균  
    1. 읽는법 : 민,'엑스바'
    2. 정의 : 엑스바는 n분(전체 개수)의 시그마(합계)는 1부터 n까지(전체 개수) xi(i번째 데이터)다.
    3. 수식 :N 또는 n은 주로 레코드나 관측값의 개수로 나타낸다. (통계학은 모집단N 혹은 표본 개수n)
    4. 예시 :  (3+5+1+2)/4=11
    5. 특징 : 평균값은 "극단값"에 영향을 받는다.
    6. 절충안 : 절사 평균, 가중 평균, 중간값
    
###### 절사 평균
    1. 읽는법: 트림 민, '엑스바 트림'
    2. 정의 : 값들의 크기 순으로 정렬한 후, 양끝에서 일정 개수의 값들을 삭제한 뒤 남은 값들.
    3. 수식 : P(기준 절사 값)이 앞부터 P+1번째, n(전체 데이터 개수)에서 P를 뺀 위치까지의 정렬된 
              x(i)값들을 시그마(총합)한 후, 전체 n값에서 2P를 뺀 값.
    4. 예시 : 
      - 국제 다이빙 대회에서 5명의 심판이 매긴 점수 중 가장 높은/가장 낮은 점수를 제외한 
        3명의 점수를 평균한 값으로 최종 성적을 매긴다. (유리한 심사 방지)
      - 극단값이 일부만 존재할 때 적합 -> 완전히 제거하지 않고, 핵심만 남김
    5. 특징 : 절사평균은 "극단값"의 영향을 제거한다.

###### 가중 평균 
    1. 읽는 법 : 웨이티드 민, '엑스바 더블유'
    2. 정의 :  각 데이터 값 xi에 사용자가 지정한 가중치 wi를 곱한 값들의 총합을 다시 가중치의 
               총합으로 나눔.
    3. 수식 : i=1부터 n(전체 개수)까지 각 데이터에 대해, 가중치 wi와 실패값 xi를 곱한 뒤 모두 
              더하고, 그 값을 i=1부터 n까지의 가중치 wi의 총합으로 나눈다. 
    4. 예시 
      - (아래)
      - 관측값 신뢰도/비율 다를 때 필수
    5. 특징 : 가중평균도 "극단값"의 영향을 제거한다.

###### 가중평균을 사용하게 된 두 가지 중요한 이유
    1. 신뢰도가 낮은 관측값에 낮은 가중치를 부여한다. (여러 센서 중 하나의 센서에는 낮은 가중치를 부여)
    2. 데이터를 수집할 때, 서로 다른 대조군에 대해서 항상 똑같은 수가 얻어지지 않는다.
       ( 데이터가 부족한 소수의 집단에 가중치를 더 부여한다. )

##### 1.3.2. 중간값과 로버스트 추정
  " 얼마나 극단값에 관여받지 않고 대표성을 보일까? "
 
###### 중간값
    1. 읽는법 : 메디안
    2. 정의 : 데이터에서 가장 가운데 위치한 값(유의어 : 제2사분위수, 50번째 백분위 수)
    3. 예시 : 평균이 왜곡됐을 때 비교용
    4. 특징 : - 평균보다 중간값이 위치 추정(특잇값)에 강하다 (로버스트하다)
              - 중간값은 결과를 왜곡할 수도 있는 극단값들의 영향을 받지 않는 
                로버스트한 위치 추정 방법이다.
              - 분포가 비대칭일 때도 대표값으로 안정적이다.
    5. 절충안 : 가중 중간값
    

###### 가중 중간값
    1. 읽는법 : 웨이티드 메디안
    2. 정의 : 데이터를 정렬한 후, 누적 가중치의 합이 전체 가중치의 50%를 초과하는 첫 번째 값
    3. 수식 : 주어진 Xn에 대한 가중치 Wn이 있을 때 데이터를 오름차순으로 정렬한 후, 
              누적 가중치의 Wj가 Wj를 2분의 1의 값보다 처음으로 커지는 지점의 값이 가중 
              중간값이다.
    4. 예시 : 데이터의 분포가 비대칭(소득, 예약일수) + 소수 그룹의 데이터 수가 부족.
    5. 특징 : 일반 중간값과 달리, 데이터에 가중치를 부여할 수 있어서 실전에 활용이 높음.
    
###### 극단값(특잇값)
    1. 읽는법 : 특잇값, 극단값, 아웃라이어
    2. 정의 : 어떤 데이터 집합에서 다른값들과 매우 멀리 떨어져 있는 값들.
    3. 예시
      - 서로 다른 단위의 값들(킬로미터-미터)일 때, 
      - 센서에서 잘못된 값이 읽힐 때는 극단 값이다.
    4. 특징 : 
        - 상황에 따라 다소 주관적일 수 있다. (북극-한국의 온도, 빌게이츠의 자산- 시민의 자산)
        - 잘못된 위치 추정을 예방하기 위해, 평균과 중간값을 같이 봐야 설득력있는 결과를 볼 수 있다

       

#### 1.3.3 대표값 해석 기준

    <대표값 종류  /  결과가 평균보다 낮으면 / 결과가 평균보다 높으면>
    1. 절사평균 
        / 극단값이 평균을 끌어올렸다. 
        / 극단값이 평균을 끌어내렸다.
    2. 중간값 
        / 데이터가 오른쪽에 치우쳐있다. (positive)
        / 데이터가 왼쪽으로 치우쳐있다. (negative)
    3. 가중평균 
       / 가중치(선정한 수치값)가 높은 집단이 평균보다 낮은 값을 가졌다. 
       / 가중치(선정한 수치값)이 높은집단이 평균보다 높은 값을 가지고 있다. 
    4. 가중중간값 
       / 가중치(선정한 수치값)이 낮은 값에 집중되어 있다. 
       / 가중치(선정한 수치값)이 높은 값에 집중되어 있다.
