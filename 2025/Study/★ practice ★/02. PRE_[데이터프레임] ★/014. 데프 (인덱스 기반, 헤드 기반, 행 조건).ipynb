구분               	loc[]	                           iloc[]
기준	                레이블(label) 기반	               정수(integer) 기반 인덱스
행 선택	            df.loc[3] → 인덱스가 '3'인 행 선택	   df.iloc[3] → 4번째 행 선택
열 선택	            df.loc[:, 'age'] → age 열 선택	   df.iloc[:, 2] → 3번째 열 선택
슬라이싱	            끝 포함 (inclusive)	               끝 제외 (exclusive, 파이썬 방식)
혼합	.               loc[행라벨, 열라벨]	               .iloc[행번호, 열번호]
