```python
import pandas as pd
import numpy as npi
```

```python
# 2차원 리스트 (2행 3열) -> 배열 -> DF 변경
# 배열과 DataFrame은 완전히 다르다.
# 배열은 열이 없다.
data = [ ['Jacky Jackson', 38, True],['minji jang',1,False] ]
matrix = np.array(data) #
```

```python
print(matrix)
print("차원 수:", matrix.ndim)        # → 2  : 2차원 = 엑셀 형식
print("모양 (shape):", matrix.shape)  # → (2, 3) : 2차원 이면서, 3개의 컬럼(열)
```

```python
result = pd.DataFrame(matrix, columns = ['name','age','driver'])
```
