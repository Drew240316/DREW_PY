```python
dataframe['age'] = dataframe['age'].astype(int)
```

```python
dataframe['age'].isnull() #null값이면 True
```

```python
dataframe['age'].isnull().sum()
```

```python
dataframe = dataframe.dropna(subset=['age'])
```

```python
dataframe['age'] = dataframe['age'].astype(int)
```

```python
dataframe.groupby('age').mean()

# TypeError: agg function failed [how->mean,dtype->object]
## age 열이 아니라 다른 열 때문이다. groupby는 데이터프레임 전체의 열에서 age기준으로 묶는 것이기 때문이다. -
```

```python
# 방법 1: 숫자형 열만 자동 선택
dataframe.groupby('age').mean(numeric_only=True)  # 추천

# 방법 2: 명시적으로 숫자열만 골라서 집계
#dataframe.groupby('age')[['fare', 'sibsp', 'parch']].mean()
```
