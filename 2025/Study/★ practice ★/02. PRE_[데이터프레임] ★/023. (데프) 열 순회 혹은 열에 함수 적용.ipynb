```python
import pandas as pd
dataframe = pd.read_csv("C://Users//wkdal//Desktop//archive//titanic.csv")
dataframe.head()
```

```python
# 처음 두 이름을 대문자로 바꾼다. : 열 원소 순회하기

for name in dataframe['name'][0:3]:
    print(name.upper())
```

```python
# 함수 정의 : 모든 열 원소에 함수 적용.

def uppercase(x):
    return x.upper()

dataframe['name'].apply(uppercase)[0:2]
# AttributeError: 'float' object has no attribute 'upper'
# -> NaN을 float취급함. NaN을 지우거나 대체해야 함.
```

```python
# NaN이 몇 개인지 확인.
print(dataframe['name'].isna().sum()) # isna : 결측값(Nan)확인하는 함수. # 변수['열'].isna()
```

```python
# NaN을 지움
dataframe = dataframe.dropna(subset=['name'])
# dropna() 결측값이 있는 행 또는 열 삭제
# 변수.dropna() 행 전체 삭제 # 변수.dropna(subset=['특정 열에 있는 nan행만 삭제'])
# 변수.dropna(axis = 1) 열 삭제
```
