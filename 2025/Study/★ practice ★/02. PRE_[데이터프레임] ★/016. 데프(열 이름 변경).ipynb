```python
# 변수.rename(colums = {'바꿀 대상':'새 이름'}).head()

dataframe.rename(columns = {'pclass' : 'Passenger Class'}).head(3)
```

```python
# 딕셔너리 키가 없을 시, 딕셔너리에 키(열)를 생성하려는 목적 -> 일반 딕셔너리 (dict)는 없는 키에 접근하면 KeyError 발생

import collections # pythonn 표준 라이브러리로, defaultdict는 없는 키에 자동으로 기본값을 지정

# 딕셔너리 생성

columns_names = collections.defaultdict(str) # defaultdict(str)은 없는 키에 접근하면 자동으로 빈 [문자_서식_포맷_정규] '' 생성

# 딕셔너리의 키 생성 (키는 dataframe의 columns에서 가져옴.)
for name in dataframe.columns:
    columns_names[name]

```
