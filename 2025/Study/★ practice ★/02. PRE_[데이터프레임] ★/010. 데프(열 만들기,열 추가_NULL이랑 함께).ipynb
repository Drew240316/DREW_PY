```python
import pandas as pd
```

```python
titanic = pd.read_csv("C://Users//wkdal//Desktop//archive//titanic.csv")
```

```python
titanic.head()
```

```python
# 열 추가

titanic['name_new'] = ['minji jang']
titanic['age_new'] = [30]
titanic['Driver_new'] = [True]
```

ValueError: Length of values (1) does not match length of index (1310)

열 추가 할 때, titanic 데이터프레임에는 1310개의 행(세로)가 있어서 나머지를 채울 수 없다

```python
# 나머지는 null값으로 채우고 1개의 행만 아래의 값으로 채움
titanic['name_new'] =  pd.Series(['minji jang'], index=[0])
titanic['age_new'] =  pd.Series(['1'], index=[0])
titanic['Driver_new'] =  pd.Series(['30'], index=[0])
```
