```python
import pandas as pd

dataframe = pd.read_csv("C://Users//wkdal//Desktop//archive//titanic.csv")

dataframe.head(3)
```

```python
# 시간으로 인덱스 만들기

time_index = pd.date_range('01/02/2017', periods = 1310, freq = '30s')
```

```python
#인덱스로 설정

dataframe.index = time_index
```

```python
dataframe.head()
```

```python
dataframe.index.name = 'Time_index'
```

```python
dataframe.head()
```

```python
# dataframe의 index이름은 Time_index이고 해당된 index의 컬럼명은 paclas...

print(dataframe.index.name)         # 'Time_index'
print(dataframe.columns[:])        # Index(['pclass', 'survived', 'name', 'sex', 'age'], dtype='object')
```

```python
# 시간 단위별로 그룹화해서 합/개수 등 계산

dataframe.resample('M').count()
```
