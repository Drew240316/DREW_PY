## 2.1 임의표본추출과 표본평향

  " 샘플 기반의 추정이나 모델링에서 데이터 품질은 데이터 양보다 더욱 중요하다. "
    
    - 데이터 과학에서 데이터 품질이란 완결성, 형식의 일관성, 깨끗함 및 각 데이터 값의 정확성을 말한다.
      - 완결성 : 누락 없이 값이 채워져 있는가?
      - 일관성 : 형식이나 단위가 일정한가?
      - 정확성 : 값이 사실과 맞는가? (잘못 기록되었는가?)
      - 청결함 : 중복/오타/비정상값 없이 정제되었는가?
    - 통계는 여기에 대표성이라는 개념을 추가한다.
      - 모집단을 제대로 대표하는가? (편향되면 결과가 왜곡된다.)
      
   " 그래서 현실에서는 데이터 정재(완결성, 일관성, 정확성)가 분석 가능하게 만들기 위해 필요하고, 
     대표성 점검(샘플 편향, 구성 확인)은 신뢰할 수 있는 결론을 만들기 위해 필요하다. " 
     
     cf. 좋은 데이터 분석가는 "이 데이터가 누구를 말해주는지까지 판단하는 사람"이다. 
    
  
#### 용어 정리 ( 표 모 엔 임 층 계 단 편 표 )

    1. 표본(sample) : 더 큰 데이터 집합으로부터 얻은 부분집합
    2. 모집단(population) : 어떤 데이터 집합을 구성하는 전체 대상 혹은 전체 집합
    3. N(n) : 모집단 (표본)의 크기
    4. 임의표본추출(임의표본, 랜덤표본추출,random sampling) : 무작위로 표본을 추출한 것
    5. 층화표본추출(층화표집) : 모집단을 층으로 나눈 뒤, 각 층에서 무작위로 표본을 추출하는 것
    6. 계층(stratum) : 공통된 특징을 가진 모집단의 동종 하위 그룹
    7. 단순임의표본(단순랜덤표본, simple ramdom sample) :
       모집단 층화 없이 임의표본추출로 얻은 표본
    8. 편향(bias) : 계통상의 오류
    9. 표본편향(sample bias) : 모집단을 잘못 대표하는 부분
 
 #### 주요 개념 ( 가-임, 위)
 
     1. 가설을 구체적으로 명시하고 임의표본추출 원칙에 따라 데이터를 수집하면 편향을 피할 수 있다.
     2. 모든 형태의 데이터 분석은 데이터 수집/분석 프로세스에서 생기는 편향의 위험성을 늘 갖고 있다.
        (데이터 마이닝에서의 모델 반복 실행, 연구 시 데이터 스누핑, 흥미로운 사건의 사후 선택 등).
 
 #### 임의표본추출과 표본편향 예시  (비임의)
     1936년 미국 알프레드 랜던vs루스벨트 대선 : 조사자 명단 
     - 리터러리 다이제스트 : 선정 방식보다 수를 늘림(상대적으로 사회경제적 지위가 높은 사람들 명단으로 조사)
     -> 표본편향 발생  ( 이유 : 비임의 방식으로 표본 추출 )
     -> 비임의는 "아무리 랜덤 표본이라고 해도, 어떤 표본도 모집단을 정확하게 대표할 수 없다는 것을 의미 "
 
 #### 자기선택과 표본 편향  (장단점)
     옐프(Yelp)와 같은 소셜 미디어 사이트는 레스토랑, 호텔, 카페 등에 대한 리뷰를 작성하는 사이트
     - 무작위로 선택되지 않아 편향이 쉬움
     - 작성자 스스로 리뷰 작성에 대한 주도권을 쥐고 있고, 자기선택편향으로 연결
     - 리뷰를 남기고자 하는 사람들은 시설에 대한 안 좋은 경험이 있거나, 해당 시설과 관련이 있는 사람들.
     - 장점) 어떤 시설을 비슷한 시설과 단순 비교
     - 단점) 상황을 정확히 파악하기는 힘듬

### 2.1.1 편향

임의표본추출로 인한 오류와 편향에 따른 오류는 신중하게 구분해서 봐야한다.
    
    1. 임의표본추출
      - 총알이 여기저기 흩어져서 맞음 ( 제 4분면에서 어떤건 위, 아래, 오른쪽 등)
      - 평균적으로 가운데 근처를 맞추려는 경향이 있음
      - 즉, 특정 방향으로 치우치지 않음
      - 이런 경우, 프로세스는 평향되지 않았지만 정확성은 떨어질 수 있음
      - 총알이 흩어져 있지만 전체적으로 가운데 주변 -> 편향은 없음
    
    2. 편향
      - 총알이 계속 오른쪽 위(제 1사분면)에만 맞는다고 가정
      - 이건 총 자체가 약간 오른쪽 위로 쏘는 버릇이 있는 것
      - 즉, 일정한 방향으로 계속 틀리는 것 = 편향 
      - 총알이 한쪽 방향으로 몰려 있음 -> 편향 있음
 
 해결 
 
     1. 임의표본추출에 따른 오류 해결 
       - 무작위로 표본을 뽑았기 때문에 발생하는 우연한 차이
       - 다시 표본을 뽑으면 다른 결과가 나올 수 있음
       - 오차의 크기(표본 평균이 모집단 평균에서 얼마나 벗어났는가)를 줄이려면 표본 수를 줄이면 됨
     
     2. 편향에 따른 오류 해결
       - 설계, 수집, 분석 과정에서 일정한 방향으로 치우친 오차
       - 특정 집단만 표본으로 뽑음, 질문 방식이 유도적임, 측정 기기가 한쪽으로 쏠림
       - 시스템적인 문제
       - 표본 수를 아무리 늘려도 편향은 그대로 남음.
       
     결국 전사적 편향인, 임의표본추출을 따라야한다.

### 2.1.2. 임의 선택

임의 선택을 위해, "임의 표본 추출"이나 "층화 표본 추출"을 사용하자.

    임의 표본 추출을 잘하기 위해서는,
    1. 접근 가능한 모집단을 적절하게 정의한다. (고객이 누군지 정의) 
    2. 표본추출 절차를 정해야한다. 
      - 실시간 거래 고객이나 웹 방문자와 같은 유동적 상황 고려.
      - 시기 등 고려(평일 오전 10시의 웹 방문자와 주말 오후 10시의 웹 방문자가 다를 수 있다.)

 2. 대량의 데이터는 언제 필요할까?

### 2.2 선택 편향

#### 용어 정리 ( 통계에서 자주 나타나는 선택평향의 유형들)
    가. 선택 편향 : "데이터 수집/표본 추출 과정"에서 특정 조건/집단만 선택되어 생기는 편향
    
    나. 데이터 스누핑 :" 데이터를 미리 보고" 가설/모델을 만들거나, 여러 테스트를 한 뒤 유리한 결과만 선택
    
      - 어떤 가설을 세우고 그것을 시험하기 위해서 잘 설계된 실험을 수행한다면, 
        그 결과에 대해 강하게 확신할 수 있다. 하지만 이런 경우는 드물다.

        (다시말해, 흥미로운 것이 나올 때까지 데이터를 너무 샅샅이 뒤진 결과가 아닐까?)

        예 ) 특정 정책 효과를 보기 좋은 시기만 뽑아 분석 -> 인위적인 효과 연출

 #### 2.1.1 평균으로의 회귀
 
1. 개념

     극단적인 관측값(예: 매우 높은 점수, 매우 낮은 수익률 등)은 다음 측정에서는 평균에 가까워질 가능성이 높다      는 통계적 경향성이다. 

2. 예시

    첫 시험에서 아주 높은 점수를 받은 학생이 다음 시험에서는 평균보다 조금 낮은 점수를 받는 경우.

    어떤 운동선수가 어느 시즌에 엄청난 성적을 냈지만 다음 시즌에는 평범한 성적으로 돌아오는 경우.

3. 주의할 점

    예외적인 결과가 관찰됐을 때, 그 자체를 너무 특별하게 해석하거나 개입의 효과로 착각하면 안된다.

    **선택 편향(selection bias)**이나 착각된 상관관계로 이어질 수 있음.

4. 실무 적용에서의 오류
    - 어떤 마케팅 캠페인 이후 성과가 떨어졌을 때, 평균 회귀 현상인지, 정말 캠페인의 문제인지 구분하지 않으면 
      잘못된 판단을 할 수 있음.

    - 의료 연구에서도, 병이 매우 심한 환자들이 치료 후 나아지는 것은 회귀 때문일 수도 있음.
    
 cf. 여기서 회귀는 돌아가다라는 뜻으로 통계적 모델링인 선형회귀와는 구분되어야 한다. 선형회귀는 예측변수와 
     결과 변수 사이의 선형적 관계를 추정하는 방법

### 2.3 통계학에서의 표본분포

#### 용어 정리
    1. 표본 통계량 : 더 큰 모집단에서 추출된 표본 데이터들로부터 얻은 측정 지표
                     (중심경향지표-평균,중앙값, 산포-IQR,분산,표준편차,상관계수, 회귀계수)
    2. 표본 분포 : 여러 표본들 혹은 재표본들로부터 얻은 표본통계량의 도수분표
    3. 표본 분포를 얻는 2가지 방법
      - 중심극한정리, 부트스트랩
    4. 데이터 분포 : 어떤 데이터 집합에서의 각 개별 값의 도수분포
    5. 표준 편차 : 모집단 내의 개별 데이터들이 평균을 기준으로 얼마나 퍼져있는지 나타내는 값
    6. 표준 오차 : 모집단에서 뽑은 대표성있는 표본 통계량(ex.평균)이 모집단의 실제값 주위에서 얼마나 
                   변동성/불확실성/분산 되있는지 나타내는 값(*표본 통계량에 따라 값이 달라짐)
    7. 정규분포 : 데이터가 평균을 중심으로 좌우 대칭을 이루어 "종 모양"을 가지는 "확률 분포"
 
#### 주요 개념
     1. 표본통계량의 도수분포는 그 해당 지표(표본 통계량)가 표본마다 다르게 나타날 수 있음을 보여준다.
     2. 부트스트랩식 방식 혹은 중심극한정리에 의존하는 공식을 통해 표본분포를 추정할 수 있다.
     3. 표준오차는 표본통계량의 변동성을 요약하는 주요 지표다.

#### 2.3.1 중심극한정리
    모집단의 분포가 정규분포가 아니더라도, 표본 크기가 충분히 크다면,표본평균들의 분포는 정규분포에 가까워진
    다. 그래서 모집단의 정규분포가 아니더라도, 중심극한정리 덕분에, t분포나 정규분포와 같은 정규근사공식을 사
    용해 통계적으로 추론(신뢰구간, 가설 검정)이 가능하다.
  
#### 2.3.2 표준오차 공식과 고려할 사항
    1. 공식 
        표준편차를 표본 수의 제곱근으로 나눈다.
        - 제곱근인 이유 ) 표본 크기가 커질수록 평균의 불안정성은 작아짐
    2. 고려할 사항
        1) 모집단에서 완전히 새로운 샘플들을 많이 수집한다.
        2) 각각의 새 샘플에 대해 통계량 (예 : 평균)을 계산한다.
        3) 2단계에서 얻은 통게량의 표준편차를 계산한다. 이것을 표준오차의 추정치로 사용한다.

### 2.4 부트스트랩

#### 용어 정리
      1. 편향 : 여러 번 추정했을 때, 그 평균 추정값이 원래 값에서 얼마나 떨어져 있는가
      2. 편차 : 개별 데이터나 추정치가 자기 평균에서 얼마나 떨어져있는가 (분산, 표준편차 계산에 사용)
      3. 배깅 :  여러 부트스트랩 표본들로부터 얻은 예측값을 모아서 결론을 만드는 것
      4. 예측 모델 : 트랜스포머, 회귀, 트리, 신경망..

#### 주요 개념
    1. 부트스트랩(데이터로부터 복원추출)은 표본통계량의 변동성을 평가하는 도구
    2. 부트스트랩은 표본분포의 수학적 근사치에 대한 엄청난 연구 없이도, 다양한 환경에서 유사한 방식으로 적용
    3. 예측 모델을 적용할 때, 여러 부트스트랩 표본들로부터 얻은 예측값을 모아서 결론을 만드는 것(배깅)이 단일          모델을 사용하는 것보다 좋다.

    1. 정의
      현재 있는 표본에서 추가적으로 표본을 복원추출하고, 각 표본에 대한 통계량과 모델을 다시 계산하는 것
      (복원 추출 : 각각의 표본을 뽑은 후 각 관측치를 다시 원래 자리로 돌려놓는다.) 
    2. 부트스트랩 알고리즘 절차
      1) 샘플 값을 하나 뽑아서 기록하고 다시 제자리 놓는다.
      2) n번 반복한다.
      3) 재표본추출된 값의 평균을 기록한다
      4) 1~3단계를 R번 반복한다.  (R번은 임의로 설정 가능) 
      5) R개의 결과를 사용하여
          a. 부트스트랩을 통해 얻은 여러 표본 평균들로부터, 그 평균들이 얼마나 변동하는지를 나타내는 
             표준오차(대표성 있는 오차)를 구하는 것
          b. 히스토그램 또는 상자그림을 그린다.
          c. 신뢰구간을 찾는다.
    3. 부트스트랩 목적
      위의 절차를 통해 표본통계량 혹은 추정한 모델 파라미터의 부트스트랩 집합을 얻게 되고, 결과적으로
      이 집합이 얼마나 변하는지를 조사할 수 있다. 
      
    4. 배깅 프로세스 활용 가능
      다변량 데이터인 분류 및 회귀트리(의사결정트리), 트랜스포머를 사용할 때, 여러 부트스트랩 샘플을 가지고 트
      리를 여러 개 만든 다음 각 트리에서 나온 예측값을 평균 내는 것에도 사용된다.
      (트랜스포머 : 같은 데이터셋에서 여러 부트스트랩 샘플 생성 → 모델 여러 번 학습 → 예측 분포 관찰
      ⇒ 예측값의 불확실성(variance) 또는 신뢰구간을 추정할 수 있음 )

```python
import pandas as pd
from sklearn.utils import resample

# CSV 파일 경로
loans_income = pd.read_csv(
    r"C:\Users\wkdal\Desktop\데이터과학을위한 파이썬\practical-statistics-for-data-scientists-master\practical-statistics-for-data-scientists-master\data\loans_income.csv"
)

# 데이터 확인
print(loans_income.head())

```

```python
#resample() 현재 표본에서 임의로 데이터를 뽑고 다시 넣고 또 뽑는 작업을 자동으로 해줍니다.
# 부트스트랩: 'x' 컬럼 기준
results = []
for _ in range(1000):
    sample = resample(loans_income['x'])  #컬럼 이름이 x임.
    results.append(sample.median())      

results = pd.Series(results)  # 리스트 → Series

# 05. 공통_[출력_반복문]★ 중앙값
original_median = loans_income['x'].median()
bias = results.mean() - original_median
stderr = results.std()

# 결과 출력
print('Bootstrap Statistics:')
print(f'original: {original_median:.3f}')
print(f'bias: {bias:.3f}')
print(f'std. error: {stderr:.3f}')
```
