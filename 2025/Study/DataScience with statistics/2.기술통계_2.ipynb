#### 1.4 변이 추정
  " 변이 추정 > 위치 추정 값 사이 > 분산, 표준 편차 . "
  " 얼마나 퍼져 있는가 = 백분위수 "

#### 주요 개념 (분표, 특, 중백)
    - 분산과 표준편차는 가장 보편적으로 널리 사용되는 변이 측정 방법이다.
    - 이들 모두 특잇값에 민감하다
    - 중간값과 백분위수(분위수)로부터 평균절대편차와 중간값의 중위절대편차를 구하는 것이
      좀 더 로버스트하다.

#### 용어 정리
    1. 편차 : 관측값과 위치 추정값 사이의 차이이며, 중앙값을 주변으로 얼마나 펴져있는지 확인하는 용도
              (유의어 : 오차, 잔차)
    2. 편차의 합은 0 :
      왜? 평균은 데이터의 '균형점'이기 때문에, 좌우의 편차가 서로 상쇄한다.
          이는 모든 관측값들과의 거리(편차)의 총합이 0이 된다.
          결국 변동성을 알 수 없다.
    3. 평균절대편차 : 데이터 값들이 평균으로부터 얼마나 떨어져 있는지를 절댓값 기준으로 평균 낸 값 
       ( 유의어 : l1 노름, 맨해튼 노름 )
    4. n-1 :
          분산은 데이터 값이 얼마나 밀집해 있고, 퍼져있는 값을 보려고 한다.
          하지만, n(표본 개수)을 가지고 N(모집단 개수)을 추정하고자 한다는 전제로 한다.
          그래서 
          - 표본 평균은 모집단 전체의 진짜 평균이 아니다.
          - 이 표본 평균은 샘플로 만들어진 값이여서 -> 보통 데이터 중심(대표)쪽으로 몰리는 경향이 있다.
          - 그래서 이 표본 평균을 기준으로 편차를 계산하면, 분산(전체 퍼진 정도)가 실제보다 작게 나올 위험이 있
            다. -> 과소평가(bias)
          - 표본 개수(n)로 나누면 과소추정이되니, 평균 1개를 잃어버려 나누는 수가 작아지니 값을 더 크게 만든다.
            (퍼짐을 넓게 봐서 일부로 더 넓은 분산을 본다.)
    5. 분산 : 데이터가 평균을 기준으로 얼마나 퍼져 있는지를 제곱한 편차의 평균으로 나타낸 값 
             (유의어 : 평균제곱오차)
    6. 표준편차 : 분산의 제곱근 
    7. 중간값의 중위절대편차(MAD) : 중간값과의 편차의 절댓값의 중간값
    8. 범위 : 데이터의 최댓값과 최솟값 차이
    9. 순서통계량 : 최소에서 최대까지 정렬된 데이터 값에 따른 계량형(유의어 : 순위)
    10. 사분위수범위(IQR) : 75번째 백분위수와 25번째 백분위수 사이의 차이
    11. 백분위수 : 데이터를 100등분했을 때,특정 값이 전체 중 몇 % 아래에 위치하는지를 나타내는 지표
    12. 도수분포표 : 변수의 범위를 동일한 크기의 구간으로 나눈 다음, 각 구간마다 몇 개의 변수값이 있는지 알
                    아보는 것
    13. 히스토그램 : 변수의 범위를 동일한 크기의 구간으로 나눈 다음, 각 구간마다 몇 개의 변수값이 있는지 알아보는 것을 막대그래프로 시각화 한 것.

#### 1.4.1 표준편차와 관련 추정값들 

  " 가장 대표적으로 사용되는변이 추정은 위치 추정값 사이의 차이다. "

##### 편차
    1. 읽는법 : 편차
    2. 정의 : 관측값과 위치 추정값 사이의 차이이며, 중앙값을 주변으로 얼마나 펴져있는지 확인하는 용도
              (유의어 : 오차, 잔차)
    3. 예시 : {1,4,4}라는 데이터가 있다고할 때, 평균은 3, 중간값은 4다.
              평균의 편차는 1-3=-2, 4-3=1, 4-3=1이다. 
    4. 특징 :음수면 평균보다 작다. 
             양수면 평균보다 크다.
             크기 얼마나 멀리떨어졌는지 확인할 수 있다. 
     

##### 평균절대편차
    1. 읽는법 : 민 앱솔루트 데비에이션
    2. 정의 : 데이터 값들이 평균으로부터 얼마나 떨어져 있는지를 절댓값 기준으로 평균 낸 값 
    3. 수식 : i = 1부터 n까지의 모든 관측값 xi에서 엑스바(평균값)을 뺀 편차의 절댓값을 
              모두 시그마(합)하고, 그 시그마 값을 n(관측값 개수)로 나눈다. 
    4. 예시 : 기상 데이터의 평균절대편차가 크다면, 날씨 변화가 심하다는 의미다.
              (2+2+1) / 3 = 1.33


##### 분산
    1. 읽는법 : 베리언스 , 에스투
    2. 정의 : 데이터가 평균을 기준으로 얼마나 퍼져 있는지(편차)를 제곱하고, 평균으로 나타낸 값
     <제곱하는 이유?>
      - 크게 벗어난 값을 제곱하면 값이 커져서 강조된다. = 평균에서 많이 떨어진 극단값을
        더 크게 반영해서 흩어짐의 심함을 직관적으로 본다.
     - 양수,음수를 상쇄한다.
    3. 수식 : i부터 n까지의 모든 관측값들의 xi에서 엑스바(표본 평균)을 뺀 값을 제곱해서 
              모두 더하고 n-1로 나눈다.
    4. 예시 : 기계 부품의 무게 분산이 작다면, 제품의 품질이 일정하다는 의미다.

##### 표준편차 ** 전체 퍼짐(이상치 포함)을 봐라

    1. 읽는법: 스탠다드 데비에이션, 에스
    2. 정의 : 데이터가 평균을 기준으로 얼마나 퍼져 있는지(편차)를 제곱하고, 
              평균으로 나타낸 값의 평균 거리(제곱근)
    3. 수식 : 분산의 제곱근(루트)
    4. 제곱근하는 이유?
      - 분산에서 퍼짐이 얼마정도로 보면, 해석 가능한 값으로 변경하기 위해 루트를 사용
        (원래 단위/값 변경)
    5. 예시 : 표준편차가 큰 투자 상품일수록, 수익률의 변동성이 크고 리스크가 크다.

##### ** 중위 절대 편차 (MAD)  : 이상치에 둔감한가? 일반적인 데이터들인가? 
    1. 읽는법: MAD 
    2. 정의 : 중간값과의 편차의 절댓값의 중간값
    3. 수식 : 각 데이터 값에서 전체 데이터의 중앙값을 뺀 뒤, 그 편차의 절댓값들을 구하고,
              그 절댓값들의 중앙값을 다시 계산한 값
    4. 예시 : 
      - 어떤 반 학생들의 키가 [160, 162, 163, 164, 165]cm일 때, 중앙값은 163cm이고,
        각 학생의 키에서 중앙값을 뺀 절댓값은 [3, 1, 0, 1, 2]가 된다.
        이 절댓값들의 중앙값, 즉 1이 바로 MAD다.
      - 공장에서 생산된 부품 길이의 MAD가 작다면,
        대부분의 부품 길이가 정해진 기준값(중앙값) 근처에 안정적으로 몰려 있다는 뜻이다.
      - 특정 주식의 일일 수익률에 대해 MAD가 크다면,
        극단적인 변동 없이도 꾸준히 넓게 흔들리는 시장이라는 것을 의미한다.

#### 1.4.2 백분위수에 기초한 추정값

  " 변이를 추정하는 또 다른 접근은 정렬된 데이터가 얼마나 퍼져 있는지를 보는 것이다. " 

##### 순서 통계량과 범위
    정렬된 데이터가 얼마나 퍼져있는지 보기 위해, 정렬(순위) 데이터를 나타내는 통계량을 순서통계량이라 하고,
    여기서 가장 기본이 되는 측도는 가장 큰 값(최댓값)과 가장 작은 값(최솟값)의 차이를 나타내는 범위다.
    최댓값, 최솟값은 특잇값에 매우 민감하여 변이를 측정하는데 유용하진 않다. 
    특잇값에 민감한 것은 피하기 위해, 사분위수(IQR)를 고려한다. 

#####  ** 사분위수(IQR) : 중간 그룹의 넓이를 봐라 
    1. 읽는법 :IQR
    2. 정의 : 75번째 백분위수와 25번째 백분위수 사이의 차이
    3. 예시 : 콜센터에서 고객 대기 시간의 3사분위수가 5분이라면,
              전체 고객의 75%는 5분 이내에 상담을 받았다는 뜻이다.
   

#### 1.4.3 세 가지 분산 지표의 해석 기준표

지표 / 수치가 클수록 / 수치가 작을 수록 

표준편차 
    / 전체 데이터가 평균으로부터 멀리 흩어짐 -> 변동성이 크다 = 극단값 영향을 받을 수 있다.
    / 대부분 데이터가 평균 주변에 모여 있다. -> 일관성 있다.

사분위범위 (IQR)
    / 중앙 50% 데이터가 넓게 퍼졌다. -> 중간 그룹 간 격차가 크다.
    / 중앙값 주변에 데이터 밀집 -> 중간층이 균일하다.
 
 중앙절대편차 (MAD)
     / 일반적인 데이터도 중간값에서 많이 벗어 났다. -> 비일관적이다.
     / 중앙값 근처에 데이터가 안정적으로 분포했다.

전체 주들 중 **중간 50%**에 해당하는 주들의 인구 수는

각 주의 인구 수가 중앙값 기준으로 약 384만 명 정도 떨어져 있다는 의미다.

이는 **극단적인 인구 수를 가진 주(예: 캘리포니아)**의 영향을 거의 받지 않고,

전체적인 일반 주들의 인구 분포를 안정적으로 요약해준다.

### 1.4.4.3 도수분포표와 히스토그램

```python
# 도수분포표
# pandas의 cut()	인구를 구간별로 나눔 (10개 구간)
# groupby(...).apply(list)	각 구간에 포함된 주들의 약자 리스트 생성
# .to_string(index=False)	전체 내용 줄임 없이 콘솔에 출력

# 1. 구간 컬럼 추가
state['binnedPopulation'] = pd.cut(state['Population'], bins=15)

# 2. 도수분포표 확인
freq_table = state['binnedPopulation'].value_counts().sort_index()
print(freq_table)
```

```python
# 3. 구간별 주 보기


## 구간 별 주의 약자 리스트 넣기 
binned_states = state.groupby('binnedPopulation')['Abbreviation'].apply(list).reset_index()

## 컬럼 이름 지정 
binned_states.columns = ['Population Bin', 'States in Bin']
print(binned_states)
```

```python
# 4. 구간별 모든 주 보기 
print(binned_states.to_string(index=False))  # 인덱스 true면 한줄씩 당겨짐
```

#### 1.4.4.4 커널 밀도 추정
  "살인율에 따른 인구 밀도"
  
    밀도 추정 
      - 히스토그램 위에 데이터를 부드러운 곡선으로 표현해, 전체 분포의 형태를 추정하는 방법

```python
# 시각화 라이브러리 다운로드
import matplotlib.pyplot as plt

# 히스토그램 + KDE(밀도 추정) 같이 그리기
mildo = state['Murder.Rate'].plot.hist(  # 'Murder.Rate' 컬럼으로 히스토그램 그리기
    density=True,                        #     👉 y축을 '빈도수'가 아니라 '밀도(확률)'로 설정
    bins=range(1, 15),                   #     👉 구간은 1~14까지 1단위 간격으로 나눔
    xlim=(0, 12),                        #     👉 x축 범위를 0~12로 설정
    alpha=0.5,                           #     👉 투명도 설정 (겹치는 그래프 잘 보이게)
    label='Histogram'                   #     👉 범례에 표시될 이름
)


# 커널 밀도 추정 곡선 추가
state['Murder.Rate'].plot.density(       #  같은 데이터에 대해 KDE(밀도 추정 곡선) 추가
    ax=mildo,                            #     👉 위에서 만든 mildo 히스토그램 위에 얹음
    label='KDE',                         #     👉 범례에 표시될 이름
    color='red'                          #     👉 곡선 색상은 빨간색
)


# 라벨 & 범례 설정
mildo.set_xlabel('Murder Rate (per 100,000)')  # 📌 x축 라벨 설정
mildo.set_ylabel('Density')                    # 📌 y축 라벨 설정
mildo.set_title('Murder Rate Distribution with Density Estimation')  # 📌 그래프 제목
mildo.legend()                                 # 📌 범례 표시 (label='...'에서 지정한 이름들)
```

#### 모멘트 해석

모멘트 차수 / 통계학적 의미 / 그래프에서의 해석 / 실무적 해석

1차 모멘트 (위치) / 중심 위치 (평균) / KDE 곡선이 2~5 사이에 중심을 둠 → 평균과 중앙값은 이 근처 / 대부분의 주는 살인율이 낮고, 그 근처에 사람들이 많이 몰려 있음

2차 모멘트 (산포) / 퍼짐 정도 (분산/표준편차) / KDE 곡선이 넓게 퍼짐, 특히 오른쪽 8~10까지 뻗음 → 분산 크고 표준편차도 큼 / 살인율이 다양한 주가 많고, 아주 높은 값도 몇 개 있어서 전체적으로 퍼져 있음

3차 모멘트 (왜도) / 비대칭성 / 오른쪽 꼬리가 더 길다 → 양의 왜도 (Positive Skewness) / 대부분 낮지만, 일부 주는 살인율이 유독 높아서 평균을 끌어올리고 있음

4차 모멘트 (첨도) / 뾰족함 / 꼬리 두께 / 정규분포보다 덜 뾰족하고 꼬리는 조금 두꺼움 → 첨도는 3보다 약간 낮거나 비슷할 수 있음 / 극단적으로 튀는 주는 많지 않고, 전체적으로는 비교적 넓게 퍼진 형태임
